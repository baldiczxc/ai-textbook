## Что такое база данных?

**База данных (БД)** — это организованная коллекция структурированной информации или данных, которая обычно хранится в электронном виде в компьютерной системе. Базы данных управляются системами управления базами данных (СУБД), которые предоставляют пользователям и приложениям средства для создания, чтения, обновления и удаления данных

database.original.jpg

Для того, чтобы понять, что такое база данных, давайте рассмотрим на примере один из разновидностей баз данных: реляционной (табличной).

Представьте, что у вас есть записная книжка, в которой вы храните контактные данные своих друзей: имена, номера телефонов, адреса и дни рождения. **Эта записная книжка - пример базы данных.**



\\

| Имя       | Номер телефона | Адрес              | День рождения |
|-----------|----------------|--------------------|---------------|
| Анна      | 123-4567       | ул. Лесная, 10     | 12.03.1990    |
| Борис     | 987-6543       | ул. Морская, 22    | 23.07.1985    |
| Виктория  | 555-1234       | ул. Центральная, 5 | 08.11.1992    |

\\


На компьютере все это происходит быстрее и удобнее c помощью баз данных. Вы можете хранить гораздо больше информации, быстро искать нужные данные, обновлять их и делать множество других операций.

## Из чего состоит база данных?

* **Таблица:**

    Вся информация хранится в таблице. В нашем примере это одна таблица, где каждая строка - это запись (данные о друге), а каждый столбец - это поле данных (имя, номер телефона, адрес и день рождения).

* **Записи (строки):**

    Каждый друг - это отдельная запись в таблице. Например, Анна - это одна запись, Борис - другая.

* **Поля (столбцы):**

    Информация о каждом друге разделена на поля. У нас есть четыре поля: Имя, Номер телефона, Адрес и День рождения.

## Что можно делать с базой данных?



Над базами данных можно выполнять так называемые CRUD (Create, Read, Update, Delete) операции:

1. **Добавление новой записи (Сreate)**:

    Вы познакомились с новым другом Петром и записали его данные в свою книжку. Это добавление новой записи в базу данных.

2. **Поиск информации (Read)**:
    
    Если вы хотите найти номер телефона Виктории, вы открываешь книжку и находите соответствующую запись. Это выполнение запроса в базе данных для поиска конкретной информации.

3. **Обновление информации (Update)**:
    
    Если Анна сменила номер телефона, вы просто исправляете его в своей книжке. В базе данных это называется обновлением записи.

4. **Удаление записи (Delete)**:

    Если вы перестали общаться с Борисом и решили удалить его данные, вы вычеркиваете его запись из книжки. В базе данных это называется удалением записи.

## Для чего нужны базы данных?

Базы данных нужны для **организации, хранения и управления большими объемами информации** таким образом, чтобы данные были легко доступны, изменяемы и защищены.

Базы данных используют:

* Банки для управления счетами клиентов, транзакциями, кредитами и т.д.
* Онлайн-магазины для хранения информации о товарах, клиентах, заказах и доставках.
* Больницы для хранения данных пациентов, медицинских записей, расписаний врачей и т.д.
* Социальные сети для хранения профилей пользователей, сообщений, фотографий и другой информации.
* и другие...

## Какие бывают базы данных?

### Реляционные базы данных (РБД)

Реляционные базы данных называются так, потому что они основаны на принципе хранения данных в таблицах и установления отношений (реляций) между этими таблицами.

На этом занятии мы познакомимся только с таблицами, но отношения между ними мы затрагивать не будем.

### NoSQL базы данных

* **Документо-ориентированные базы данных**

    Данные хранятся в виде документов, похожих на JSON.

    Пример: Папка на компьютере, где каждый документ — это запись об ученике. Каждый документ может содержать разные поля, например, у одного ученика есть телефон, а у другого — нет.

* **Базы данных "Ключ-значение"**

    Данные хранятся в виде так называемых пар "ключ-значение".

    Пример: Словарь, где ключ — это слово, а значение — его определение. Например, ключ "Анна" и значение "123-4567" (номер телефона).

* **Графовые базы данных**:

    Данные хранятся в виде узлов и связей между ними.

    Пример: Граф друзей, где каждый человек — это узел, а дружба между ними — это линия.

### Базы данных временных рядов

Оптимизированы для хранения и обработки данных, которые изменяются со временем.

Пример временных рядов: Список значений температуры и время, когда они были записаны.

### Объектно-ориентированные базы данных

Данные хранятся в виде объектов.

Пример объекта: Ученик в школе — это объект со своими свойствами и методами, как в программировании.

## Structure Query Language (SQL)

Реляционные базы данных понимают, что от них хотят, если команды для них написаны на языке SQL.

SQL (Structured Query Language, или язык структурированных запросов) — это язык, созданный специально для работы с базами данных. Он позволяет выполнять различные операции с данными, хранящимися в базах данных, такие как создание, чтение, обновление и удаление данных.

Вот основные операции, которые можно выполнять с помощью SQL:

1. **SELECT**: используется для выборки данных из базы данных. Например, можно выбрать все книги, написанные определённым автором.
    ```sql
    SELECT * FROM books WHERE author = 'Пушкин';
    ```

2. **INSERT**: используется для добавления новых записей в таблицу.
    ```sql
    INSERT INTO books (title, author, year) VALUES ('Евгений Онегин', 'Пушкин', 1833);
    ```

3. **UPDATE**: используется для обновления существующих данных в таблице.
    ```sql
    UPDATE books SET year = 1837 WHERE title = 'Евгений Онегин';
    ```

4. **DELETE**: используется для удаления записей из таблицы.
    ```sql
    DELETE FROM books WHERE title = 'Евгений Онегин';
    ```

PostgreSQL (также называемая Postgres) — это мощная, открытая объектно-реляционная система управления базами данных (СУБД). Она известна своей расширяемостью (через пользовательские функции) и соблюдением стандартов SQL, что делает её популярной среди разработчиков и предприятий. Эта СУБД обеспечивает надежность и стабильность тем, что позволяет восстанавливать данные после сбоев. Активное сообщество и регулярные обновления способствуют её постоянному улучшению и внедрению новых возможностей.

### Внутренная структура Postgres

03fig02.jpg

1. **Кластер (Cluster)**

    Это самая верхняя часть структуры. Кластер включает в себя несколько баз данных (databases) и отвечает за управление ими. Кластер создается, когда устанавливается PostgreSQL, и все базы данных в системе будут находиться в этом кластере.

2. **Базы данных (Databases)**

    База данных — это контейнер для данных. В одном кластере PostgreSQL может быть несколько баз данных, но каждая база данных изолирована и не может напрямую обращаться к данным другой базы данных.

3. **Схемы (Schemas)**

    Схема — это логический контейнер внутри базы данных. Она содержит таблицы и другие объекты базы данных. Схемы помогают организовать данные и управлять доступом к ним. В одной базе данных может быть несколько схем.

4. **Таблицы (Tables)**

    Таблицы — это структуры, где непосредственно хранятся данные. Каждая таблица состоит из строк и столбцов, где строки представляют отдельные записи, а столбцы — отдельные поля данных.


**Для того, чтобы лучше понять структуру Postgres, давайте рассмотрим такой пример:**

PostgreSQL — это большая библиотека (кластер), где

* Кластер — это сама библиотека, здание, где хранятся книги.
* Базы данных — это отдельные залы в библиотеке, каждый зал содержит свою коллекцию книг.
* Схемы — это книжные шкафы в зале, каждый шкаф содержит определенную категорию книг.
* Таблицы — это полки в шкафу, на которых лежат книги (данные).

### Модуль psycopg2

Psycopg2 — это библиотека для языка программирования Python, которая позволяет работать с базами данных PostgreSQL в Python. Она обеспечивает интерфейс для подключения к базе данных, выполнения SQL-запросов и получения результатов.

Через эту библиотеку мы и будем взаимодействовать с Postgres.

### Создание базы данных

Здесь мы создаем новую базу данных `mydatabase` в системе управления базами данных PostgreSQL.

Вот что она делает по шагам:

1. `PGPASSWORD='postgres'` - указывает пароль для пользователя базы данных. В данном случае, пароль - 'postgres'.
2. `psql` - это программа командной строки для работы с PostgreSQL.
3. `-U postgres` - указывает имя пользователя базы данных, который будет выполнять команду. В данном случае, пользователь - 'postgres'.
4. `-h localhost` - указывает сервер базы данных, к которому подключается. 'localhost' означает, что подключение происходит к серверу на этом же компьютере.
5. `-c "CREATE DATABASE mydatabase;"` - выполняет команду SQL для создания новой базы данных с именем 'mydatabase'.

````python
!PGPASSWORD='postgres' psql -U postgres -h localhost -c "CREATE DATABASE mydatabase;"
````

### Подключение к базе

Для того, чтобы начать работу с новой базой `mydatabase`, нужно подключиться к ней. В Python это можно сделать, используя psycopg2 вот таким образом:

````python
import psycopg2  # Импортируем библиотеку psycopg2, которая позволяет работать с базой данных PostgreSQL
from psycopg2 import sql  # Импортируем модуль sql из psycopg2 для работы с SQL-запросами

# Устанавливаем соединение с базой данных PostgreSQL
conn = psycopg2.connect(
    dbname="mydatabase",  # Имя базы данных, которую мы создали
    user="postgres",      # Имя пользователя
    password="postgres",  # Пароль пользователя
    host="localhost",     # Хост (сервер), где находится база данных. 'localhost' означает, что база находится на этом же компьютере
    port="5432"           # Порт — это своеобразный "дверной номер" на компьютере, который используется для отправки и получения данных между программами или устройствами в сети.
)

# Создаем объект cursor, который позволит нам выполнять SQL-запросы к базе данных
cur = conn.cursor()
````

### Создание таблицы (Create) вставка данных

Хорошо, мы подключились к базе. Но она сейчас пуста. Для того, чтобы наполнить её информацией, создадим таблицу `employees`, где будем хранить информацию о сотрудниках одной известной компании.

````python
# Это SQL-запрос для создания новой таблицы в базе данных
create_table_query = '''
CREATE TABLE IF NOT EXISTS employees (      -- Создаем таблицу с именем 'employees', если не была создана ранее
    id SERIAL PRIMARY KEY,                  -- Столбец 'id' будет уникальным идентификатором для каждого сотрудника и автоматически увеличиваться
    name VARCHAR(100),                      -- Столбец 'name' для хранения имени сотрудника, длиной до 100 символов
    department VARCHAR(50),                 -- Столбец 'department' для хранения отдела, в котором работает сотрудник, длиной до 50 символов
    salary NUMERIC                          -- Столбец 'salary' для хранения зарплаты сотрудника, может быть любым числом
)
'''

# Выполняем SQL-запрос для создания таблицы
cur.execute(create_table_query)

# Сохраняем изменения в базе данных
conn.commit()
````

Мы создали таблицу, но она пока тоже пуста. Давайте добавим данные об одном из сотрудников:

````python
# Это SQL-запрос для добавления нового сотрудника в таблицу 'employees'
insert_query = '''
INSERT INTO employees (name, department, salary)    -- Указываем, что будем вставлять данные в столбцы 'name', 'department' и 'salary'
VALUES (%s, %s, %s)                                 -- Места для данных, которые будут вставлены. s с процентом — это заполнители для значений
'''

# Данные сотрудника, которые мы хотим добавить
employee_data = ('John Doe', 'Engineering', 60000)

# Выполняем SQL-запрос и передаем данные сотрудника
cur.execute(insert_query, employee_data)

# Сохраняем изменения в базе данных, чтобы новый сотрудник был добавлен
conn.commit()
````

### Чтение данных (Read)

Мы заполнили базу информацией. Теперь, давайте посмотрим какую информацию мы туда загрузили:  

````python
def show_employees():

    # Это SQL-запрос для получения всех данных из таблицы 'employees'
    select_query = 'SELECT * FROM employees'

    # Выполняем SQL-запрос, чтобы получить данные из таблицы
    cur.execute(select_query)

    # Получаем все строки, которые вернул запрос
    rows = cur.fetchall()

    # Проходим по каждой строке в списке строк
    for row in rows:
        # Выводим строку на экран
        print(row)

show_employees()
````

### Обновление данных (Update)

Сотрудник по имени John Doe продвинулся вверх по карьерной лестнице и стал зарабатывать больше. Давайте обновим информацию о его зарплате:

````python
# Это SQL-запрос для обновления зарплаты сотрудника в таблице 'employees'
update_query = '''
UPDATE employees    -- Указываем, что будем обновлять данные в таблице 'employees'
SET salary = %s     -- Указываем, что будем изменять зарплату. s с процентом — это место для нового значения зарплаты
WHERE name = %s     -- Указываем условие: обновляем зарплату только для сотрудника с указанным именем
'''

# Новая зарплата, которую мы хотим установить
new_salary = 65000

# Имя сотрудника, для которого нужно обновить зарплату
employee_name = 'John Doe'

# Выполняем SQL-запрос и передаем новые данные: зарплату и имя сотрудника
cur.execute(update_query, (new_salary, employee_name))

# Сохраняем изменения в базе данных, чтобы обновленная зарплата была записана
conn.commit()
````

Теперь посмотрим какая информация хранится в базе:

````python
show_employees()
````

### Удаление данных (Delete)

John Doe проработал в компании много лет, и теперь хочет что-то поменять в своей жизни. Он решил уволиться из компании. Давайте удалим его из нашей базы:

````python
# Это SQL-запрос для удаления сотрудника из таблицы 'employees'
delete_query = '''
DELETE FROM employees   -- Указываем, что будем удалять данные из таблицы 'employees'
WHERE name = %s         -- Указываем условие: удаляем только того сотрудника, чье имя совпадает с указанным
'''

# Имя сотрудника, которого мы хотим удалить из базы данных
employee_name_to_delete = 'John Doe'

# Выполняем SQL-запрос и передаем имя сотрудника, которого нужно удалить
cur.execute(delete_query, (employee_name_to_delete,))

# Сохраняем изменения в базе данных, чтобы удаление было окончательным
conn.commit()
````

Убедимся, что данных о John Doe в базе теперь нет:

````python
show_employees()
````

### Отключение от базы данных

Мы закончили работать над базой данных. Как и в реальной жизни, мы не можем уйти не закрыв дверь, поэтому давайте закроем курсор и соединение:

````python
# Закрываем курсор и соединение
cur.close()
conn.close()
````
