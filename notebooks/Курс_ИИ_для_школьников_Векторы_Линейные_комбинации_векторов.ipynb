## Векторы. Линейные комбинации векторов

### Что такое вектор

*Вектор* - это упорядоченный набор чисел (называемых *компонентами вектора*), который можно записать вот таким вот образом:

$$
\mathbf{v} = \begin{pmatrix} v_1 \\ v_2 \end{pmatrix}
\text{ или } \mathbf{v} = (v_1, v_2)
$$
где
* $v_1$ - первая компонента вектора
* $v_2$ — вторая компонента вектора

Обратите внимание, что вектор может состоять не только из 2 компонент, но и из любого количества компонент.

Компоненты вектора можно воспринимать как координаты. Тогда, визуально (геометрически), вектор можно представить как стрелку, направленную из начала координат в точку с координатами, заданными компонентами вектора.

Пусть вектор $\mathbf{v}$ будет задан так:
$$\mathbf{v} = \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} 3 \\ 4 \end{pmatrix}$$

тогда на плоскости вектор будет выглядеть вот так:

```python
import matplotlib.pyplot as plt
import numpy as np

# Исходные данные для вектора
vector = np.array([3, 4])
origin = np.array([0, 0])  # Начало координат

# Создание графика
plt.figure(figsize=(6, 6))
plt.quiver(*origin, *vector, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-1, 5)
plt.ylim(-1, 5)

# Настройка осей
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Добавление меток
plt.text(vector[0] / 2 - 0.1, vector[1] / 2 + 0.1, 'v', fontsize=15, ha='right')

# Заголовок и метки осей
plt.title("Вектор на плоскости")
plt.xlabel("x")
plt.ylabel("y")

# Показ графика
plt.show()
```

### Сложение векторов

Вектора можно складывать друг с другом, чтобы получить новые вектора. Сложение осуществляется покомпонентно: это значит, что чтобы сложить вектора, нужно сложить числа из всех векторов в каждой компоненте.

Например, у нас есть вектор $\mathbf{u}$, заданный как
$$
\mathbf{u} = \begin{pmatrix} u_1 \\ u_2 \end{pmatrix}
$$

и вектор $\mathbf{v}$, заданный как

$$
\mathbf{v} = \begin{pmatrix} v_1 \\ v_2 \end{pmatrix}
$$

тогда сложение $\mathbf{u} + \mathbf{v}$ двух векторов будет

$$
\mathbf{u} + \mathbf{v} = \begin{pmatrix} u_1 \\ u_2 \end{pmatrix} + \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \begin{pmatrix} v_1 + u_1 \\ v_2 + u_2 \end{pmatrix}
$$

Теперь посмотрим это на примере. Возьмем вектор $\mathbf{u} = (3, 0)$ и вектор $\mathbf{u} = (0, 4)$. Тогда их сложение будет

$$
\mathbf{u} + \mathbf{v} = \begin{pmatrix} 3 \\ 0 \end{pmatrix} + \begin{pmatrix} 0 \\ 4 \end{pmatrix} =
\begin{pmatrix} 3 + 0 \\ 0 + 4 \end{pmatrix} =
\begin{pmatrix} 3 \\ 4 \end{pmatrix}
$$

Посмотрим как сложение векторов будет выглядеть визуально на плоскости. Чтобы получить вектор как результат сложения, нужно построить один вектор от начала координат, затем построить второй вектор от конца первого вектора. Будет это выглядеть вот таким образом:

```python
import matplotlib.pyplot as plt
import numpy as np

# Исходные данные для вектора
vector = np.array([3, 4])
u = np.array([3, 0])
v = np.array([0, 4])
origin = np.array([0, 0])  # Начало координат

# Создание графика
plt.figure(figsize=(6, 6))
plt.quiver(*origin, *vector, angles='xy', scale_units='xy', scale=1, color='r')

# Векторы u и v
plt.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(*u, *v, angles='xy', scale_units='xy', scale=1, color='b')

# Настройка осей
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Добавление меток
plt.text(vector[0] / 2, vector[1] / 2 + 0.1, f'u+v = {tuple(vector)}', fontsize=10, ha='right')
plt.text(vector[0] / 2, -0.2, f'u = {tuple(u)}', color='g', fontsize=10, ha='center')
plt.text(u[0] + 0.2, vector[1] / 2, f'v = {tuple(v)}', color='b', fontsize=10)
# Заголовок и метки осей
plt.title("Сложение векторов")
plt.xlabel("x")
plt.ylabel("y")

# Показ графика
plt.show()
```

### Умножение вектора на скаляр

Вектор можно умножать на число, называемое *скаляром*. При умножении вектора на число (скаляр) каждая его компонента умножается на это число. Если $c$ — скаляр и $\mathbf{v} = (v_1, v_2)$, то:

$$
c \cdot \mathbf{v} =
c \cdot \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} =
\begin{pmatrix} c \cdot v_1 \\ c \cdot v_2 \end{pmatrix}
$$


Геометрически, скаляр может либо увеличивать вектор по длине, либо уменьшать, либо напрявлять в противоположную сторону

```python
import matplotlib.pyplot as plt
import numpy as np

# Определение исходного вектора
vector = np.array([3, 4])

# Определение скаляров для трех случаев
scalars = [2, 0.5, -1]

# Цвета для векторов
colors = ['k', 'r', 'g', 'b']  # Черный для оригинального, остальные для масштабированных

# Создание фигуры и подграфиков
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Названия для графиков
titles = ['v', '2 * v', '0.5 * v', '-1 * v']

lim = 10

# Функция для отображения осей
def draw_axes(ax):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

# Визуализация оригинального вектора
draw_axes(axs[0, 0])
axs[0, 0].quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color=colors[0])
axs[0, 0].set_xlim(-lim, lim)
axs[0, 0].set_ylim(-lim, lim)
axs[0, 0].set_title(titles[0])
axs[0, 0].grid(True)

# Визуализация увеличенного вектора (умножение на 2)
draw_axes(axs[0, 1])
scaled_vector = scalars[0] * vector
axs[0, 1].quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color=colors[1])
axs[0, 1].set_xlim(-lim, lim)
axs[0, 1].set_ylim(-lim, lim)
axs[0, 1].set_title(titles[1])
axs[0, 1].grid(True)

# Визуализация уменьшенного вектора (умножение на 0.5)
draw_axes(axs[1, 0])
scaled_vector = scalars[1] * vector
axs[1, 0].quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color=colors[2])
axs[1, 0].set_xlim(-lim, lim)
axs[1, 0].set_ylim(-lim, lim)
axs[1, 0].set_title(titles[2])
axs[1, 0].grid(True)

# Визуализация вектора, направленного в противоположную сторону (умножение на -1)
draw_axes(axs[1, 1])
scaled_vector = scalars[2] * vector
axs[1, 1].quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color=colors[3])
axs[1, 1].set_xlim(-lim, lim)
axs[1, 1].set_ylim(-lim, lim)
axs[1, 1].set_title(titles[3])
axs[1, 1].grid(True)

# Подписи осей
for ax in axs.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('y')

# Настройка интервала между подграфиками
plt.tight_layout()

# Отображение фигуры
plt.show()
```

### Нулевой вектор

Так как мы можем умножить вектор на любое число, мы также можем умножить его на 0. В таком случае мы получаем **нулевой вектор**.

**Нулевой вектор** — это особый вектор в векторном пространстве, у которого все компоненты равны нулю. Он обозначается как $\mathbf{0}$ и имеет вид:

$$
\mathbf{0} = [0, 0, \dots, 0]
$$

В двумерном пространстве $\mathbb{R}^2$ это будет вектор $\mathbf{0} = [0, 0]$, в трёхмерном $\mathbb{R}^3$ — $\mathbf{0} = [0, 0, 0]$, и так далее.

Вектор не отображается как стрелка, поскольку его длина равна нулю, и он совпадает с началом координат. Нулевой вектор не имеет направления и длины, что делает его особым элементом в любом векторном пространстве.

```python
import matplotlib.pyplot as plt
import numpy as np

# Определим нулевой вектор
zero_vector = np.array([0, 0])

# Создадим фигуру и оси
fig, ax = plt.subplots()

# Визуализируем нулевой вектор (в этом случае просто начало координат)
ax.quiver(0, 0, zero_vector[0], zero_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Zero Vector')

# Настройки осей
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True)

# Добавляем легенду
ax.legend()

# Показать график
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Линейная комбинация векторов

Комбинируя сложение с умножением векторя на скаляр мы получем *линейную комбинацию* векторов: умножаем $\mathbf{v}$ и $c$, уможаем $\mathbf{u}$ и $d$, затем складываем
$$ c \cdot \mathbf{v} +  d \cdot \mathbf{u} $$

Таким образом, сумма $c \cdot \mathbf{v}$ и $d \cdot \mathbf{u}$ называется линейной комбинацией векторов $\mathbf{v}$ и $\mathbf{u}$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Определение векторов v и u
v = np.array([4, 2])
u = np.array([-1, 2])

# Определение выражений для каждого случая
expressions = [
    v + u,  # 1v + 1u
    v - u,  # 1v - 1u
    0*v + 0*u,  # 0v + 0u
    2*v + 0*u   # 2v + 0u
]

# Названия для графиков
titles = [
    '1v + 1u',
    '1v - 1u',
    '0v + 0u',
    '2v + 0u'
]

# Цвета для векторов
colors = ['b', 'r', 'g', 'm']

# Создание фигуры и подграфиков
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

lim = 15

# Функция для отображения осей
def draw_axes(ax):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

# Визуализация векторов на каждом подграфике
for i, ax in enumerate(axs.flat):
    draw_axes(ax)

    # Вектор v
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='k', label='v')
    ax.annotate('v', (v[0], v[1]), textcoords="offset points", xytext=(10,-10), ha='center')

    # Вектор u
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='c', label='u')
    ax.annotate('u', (u[0], u[1]), textcoords="offset points", xytext=(-10,10), ha='center')

    # Результирующий вектор
    ax.quiver(0, 0, expressions[i][0], expressions[i][1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=titles[i])
    ax.annotate(titles[i], (expressions[i][0], expressions[i][1]), textcoords="offset points", xytext=(10,0), ha='left', va='center')

    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)
    ax.set_title(titles[i])
    ax.grid(True)

# Подписи осей
for ax in axs.flat:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

# Настройка интервала между подграфиками
plt.tight_layout()

# Отображение фигуры
plt.show()
```

### Скалярное (внутреннее) произведение векторов

Для двух векторов $\mathbf{a}$ и $\mathbf{b}$ скалярное (или внутреннее) произведение векторов вычисляется как сумма произведений соответствующих компонент

$$
\mathbf{a} \cdot \mathbf{b} = a_1 b_1 + a_2 b_2 + \dots + a_n b_n = \sum_{i=1}^{n} a_i b_i
$$

**Пример**

Рассмотрим два вектора $\mathbf{a} = (a_1, a_2)$ и $\mathbf{b} = (b_1, b_2)$ в двухмерном пространстве $ \mathbb{R}^2 $. Тогда их скалярное произведение будет:

$$
\mathbf{a} \cdot \mathbf{b} = a_1 \cdot b_1 + a_2 \cdot b_2
$$

**Свойства скалярного произведения векторов**

1. **Коммутативность**: $\mathbf{a} \cdot \mathbf{b} = \mathbf{b} \cdot \mathbf{a}$
2. **Дистрибутивность** относительно сложения: $\mathbf{a} \cdot (\mathbf{b} + \mathbf{c}) = \mathbf{a} \cdot \mathbf{b} + \mathbf{a} \cdot \mathbf{c}$
3. **Ассоциативность** относительно умножения на скаляр: $k(\mathbf{a} \cdot \mathbf{b}) = (k\mathbf{a}) \cdot \mathbf{b} = \mathbf{a} \cdot (k\mathbf{b})$

### Длина вектора

Что мы получим, если посчитаем скалярное произведение для одного и того же вектора (в случае когда $\mathbf{a} = \mathbf{b}$)?

Оказывается, что в таком случае мы получаем длину вектора, возведенную во вторую степень:

$$ \mathbf{v} \cdot \mathbf{v} = v_1 v_1 + v_2 v_2 + \dots + v_n v_n = \|\mathbf{v}\|^2 $$

Таким образом, чтобы получить длину вектора, необходимо вычислить его скалярное произведение с самим собой, затем из получившегося числа извлечь корень:

$$
\|\mathbf{v}\| = \sqrt{ \|\mathbf{v}\|^2 }= \sqrt{ \mathbf{v} \cdot \mathbf{v} }  = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}
$$

Длину вектора ещё называют *нормой* вектора или *модулем* вектора

**Пример в двумерном пространстве**

Рассмотрим вектор $\mathbf{v} = (3, 4)$ в $ \mathbb{R}^2 $. Его длина будет:

$$
\|\mathbf{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

**Пример в трёхмерном пространстве**

Для вектора $\mathbf{v} = (1, 2, 2)$ в $ \mathbb{R}^3 $ его длина будет:

$$
\|\mathbf{v}\| = \sqrt{1^2 + 2^2 + 2^2} = \sqrt{1 + 4 + 4} = \sqrt{9} = 3
$$

### Угол между векторами. Перпендикулярные вектора

Скалярное прозведение векторов также можно вычислить по такой формуле:

$$ \mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos \theta $$

где:
- $\|\mathbf{a}\|$ и $\|\mathbf{b}\|$ — длины (модули) векторов $\mathbf{a}$ и $\mathbf{b}$,
- $\theta$ — угол между векторами $\mathbf{a}$ и $\mathbf{b}$,
- $\cos \theta$ — косинус угла между этими векторами

Мы можем использовать эту формулу, чтобы вычислить угол между векторами. Для этого мы сначала вычисляем значения косинуса угла между векторами:

$$  \cos \theta = \frac{ \mathbf{a} \cdot \mathbf{b} } {\|\mathbf{a}\| \|\mathbf{b}\|} = \frac{a_1 b_1 + a_2 b_2 + \dots + a_n b_n}{(a_1^2 + a_2^2 + \dots + a_n^2)(b_1^2 + b_2^2 + \dots + b_n^2)} $$

Вычислив значение косинуса, мы можем вычислить угол:

$$ \theta = \arccos ( \cos \theta ) $$

**Пример**

Рассмотрим два перпендикулярных (ортогональны) вектора на плоскости:
- $\mathbf{a} = (1, 0)$
- $\mathbf{b} = (0, 1)$

Эти векторы перпендикулярны, то есть угол между ними равен 90°.

1. Вычислим их скалярное призведение

$$
\mathbf{a} \cdot \mathbf{b} = (1 \cdot 0) + (0 \cdot 1) = 0
$$

2. Вычислим длины этих векторов

$$
\|\mathbf{a}\| = \sqrt{1^2 + 0^2} = \sqrt{1} = 1
$$
$$
\|\mathbf{b}\| = \sqrt{0^2 + 1^2} = \sqrt{1} = 1
$$

3. Используем формулу для косинуса угла

$$
\cos \theta = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|} = \frac{0}{1 \cdot 1} = 0
$$

4. Нахождение угла

$$
\theta = \arccos(0) = 90^\circ
$$

Таким образом, если скалярное произведение двух векторов равно нулю, это означает, что угол между ними равен $90^\circ$, и они перпендикулярны.

Визуализируем перпендикулярные вектора и посчитаем их скалярное произведение с помощью `np.dot`, чтобы убедиться в том, что они перпендикулярны и их скалярное произведение равно нулю:

```python
import matplotlib.pyplot as plt
import numpy as np

# Определим стандартный базис на плоскости
e1 = np.array([1, 0])
e2 = np.array([0, 1])

print("Вектор e1:", e1)
print("Вектор e2:", e2)
print("Скалярное произведение:", np.dot(e1, e2))

# Создадим фигуру и оси
fig, ax = plt.subplots()

# Визуализируем базисные векторы
ax.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r', label='e1 = [1, 0]')
ax.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='g', label='e2 = [0, 1]')

# Настройки осей
ax.set_xlim(-2, 3)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(True)

# Добавляем легенду
ax.legend()

# Показать график
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Базис

Базис в линейной алгебре — это набор векторов, которые одновременно:

1. Перпендикулярны друг другу
2. Позволяют представить любой вектор в пространстве как линейную комбинацию

Базис позволяет определить координаты вектора в векторном пространстве, что упрощает работу с векторами, их операции и преобразования.

**Пример базиса в двумерном пространстве**

В двумерном векторном пространстве стандартным базисом является пара векторов $e_1 = (1, 0)$ и $e_2 = (0, 1)$. Любой вектор $v = (x, y)$ в этом пространстве может быть представлен как линейная комбинация этих базисных векторов:
   $$
   v = x \cdot e_1 + y \cdot e_2
   $$
   где $x$ и $y$ — координаты вектора $v$ относительно базиса $\{e_1, e_2\}$.

Посмотрим как мы можем представить два разных вектра с помощью базисов $e_1$ и $e_2$:

```python
import matplotlib.pyplot as plt
import numpy as np

# Определим стандартный базис на плоскости
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Создадим пару произвольных векторов, которые являются линейными комбинациями базисных векторов
v1 = 2 * e1 + 1 * e2  # Вектор (2, 1)
v2 = -1 * e1 + 3 * e2  # Вектор (-1, 3)

# Создадим фигуру и оси
fig, ax = plt.subplots()

# Визуализируем базисные векторы
ax.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r', label='e1 = [1, 0]')
ax.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='g', label='e2 = [0, 1]')

# Визуализируем векторы v1 и v2
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1 = 2*e1 + 1*e2 = [2, 1]')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='v2 = -1*e1 + 3*e2 = [-1, 3]')

# Настройки осей
ax.set_xlim(-2, 3)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(True)

# Добавляем легенду
ax.legend()

# Показать график
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

Здесь мы можем видеть, что мы можем использовать координаты (компоненты) вектора, чтобы составить линейную комбинацию из базисных векторов и построить вектор на плоскости
