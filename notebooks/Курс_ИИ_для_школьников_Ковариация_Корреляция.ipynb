## Ковариация

Ковариация — это мера, которая показывает степень, в которой две случайные величины изменяются вместе. Она помогает понять, существует ли связь между изменениями двух величин, и если существует, то какая это связь — положительная или отрицательная.

### Определение ковариации

Ковариация между двумя случайными величинами $X$ и $Y$ обозначается как $\text{Cov}(X, Y)$ и определяется следующим образом:

$$
\text{Cov}(X, Y) = \mathbb{E}[(X - \mathbb{E}[X])(Y - \mathbb{E}[Y])]
$$

где $\mathbb{E}[X]$ и $\mathbb{E}[Y]$ — математические ожидания (средние значения) величин $X$ и $Y$ соответственно.

### Вычисление ковариации



Для выборки ковариация может быть рассчитана по следующей формуле:

$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})
$$

где:
- $n$ — количество наблюдений,
- $x_i$ и $y_i$ — значения величин $X$ и $Y$ в $i$-й выборке,
- $\bar{x}$ и $\bar{y}$ — средние значения величин $X$ и $Y$ соответственно.



### Свойства ковариации



1. Если $\text{Cov}(X, Y) > 0$, то $X$ и $Y$ имеют положительную линейную связь: когда $X$ увеличивается, $Y$ также имеет тенденцию увеличиваться, и наоборот.
2. Если $\text{Cov}(X, Y) < 0$, то $X$ и $Y$ имеют отрицательную линейную связь: когда $X$ увеличивается, $Y$ имеет тенденцию уменьшаться, и наоборот.
3. Если $\text{Cov}(X, Y) = 0$, то $X$ и $Y$ не имеют линейной связи. Однако это не означает, что они независимы, так как могут существовать нелинейные зависимости.



### Недостатки ковариации

1. **Зависимость от единиц измерения**: Ковариация зависит от единиц измерения величин. Например, если измерить одну из величин в других единицах, значение ковариации изменится.
2. **Неинформативность о степени связи**: Ковариация сама по себе не дает информации о силе линейной связи между величинами, так как ее значение зависит от масштаба данных.

## Нормализованная ковариация: коэффициент корреляции Пирсона



Для устранения недостатков ковариации часто используется нормализованная мера — коэффициент корреляции Пирсона.

### Определение

Коэффициент корреляции Пирсона обозначается как $\rho_{X,Y}$ (в популяции) или $r$ (в выборке) и определяется как нормализованная ковариация двух случайных величин $X$ и $Y$:

$$
\rho_{X,Y} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$

где:
- $\text{Cov}(X, Y)$ — ковариация между $X$ и $Y$,
- $\sigma_X$ и $\sigma_Y$ — стандартные отклонения $X$ и $Y$ соответственно.



### Вычисление корреляции

Для выборки из $n$ наблюдений формула выглядит так:

$$
r = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^n (y_i - \bar{y})^2}}
$$

где:
- $x_i$ и $y_i$ — значения переменных $X$ и $Y$ в $i$-й выборке,
- $\bar{x}$ и $\bar{y}$ — средние значения $X$ и $Y$ соответственно.

### Свойства коэффициента корреляции Пирсона



1. **Диапазон значений**: Коэффициент корреляции Пирсона принимает значения от -1 до 1:
   - $\rho_{X,Y} = 1$ означает идеальную положительную линейную связь,
   - $\rho_{X,Y} = -1$ означает идеальную отрицательную линейную связь,
   - $\rho_{X,Y} = 0$ означает отсутствие линейной связи.

2. **Безразмерность**: Коэффициент корреляции не зависит от единиц измерения величин $X$ и $Y$, что делает его удобным для сравнения связей между разными наборами данных.

3. **Симметрия**: $\rho_{X,Y} = \rho_{Y,X}$, что означает, что корреляция между $X$ и $Y$ одинакова в обоих направлениях.

### Интерпретация значений коэффициента корреляции

- **0.9 ≤ |r| ≤ 1.0**: Очень сильная линейная связь.
- **0.7 ≤ |r| < 0.9**: Сильная линейная связь.
- **0.5 ≤ |r| < 0.7**: Средняя линейная связь.
- **0.3 ≤ |r| < 0.5**: Слабая линейная связь.
- **0 ≤ |r| < 0.3**: Очень слабая или отсутствующая линейная связь.

### Примеры



1. **Положительная корреляция**: Когда значения одной переменной увеличиваются, значения другой переменной также имеют тенденцию увеличиваться. Пример: рост и вес человека (обычно наблюдается положительная корреляция).

2. **Отрицательная корреляция**: Когда значения одной переменной увеличиваются, значения другой переменной имеют тенденцию уменьшаться. Пример: скорость машины и время, необходимое для преодоления фиксированного расстояния (обычно наблюдается отрицательная корреляция).

3. **Отсутствие корреляции**: Когда изменения одной переменной не связаны с изменениями другой переменной. Пример: размер обуви и интеллект человека (нет очевидной линейной связи).



### Ограничения коэффициента корреляции Пирсона



1. **Линейность**: Коэффициент корреляции Пирсона измеряет только линейные связи. Он может быть низким, даже если между переменными существует сильная нелинейная связь.

2. **Чувствительность к выбросам**: Выбросы могут значительно повлиять на значение коэффициента корреляции, исказив результаты анализа.

3. **Не различает причину и следствие**: Высокая корреляция не означает причинно-следственную связь между переменными. Для установления причинно-следственной связи необходимы дополнительные исследования и методы.

## Корреляционная матрица

Корреляционная матрица — это таблица, которая показывает коэффициенты корреляции для каждого возможного сочетания пар случайных величин в наборе данных. Она используется для анализа взаимосвязей между множеством переменных одновременно.



### Определение

Корреляционная матрица для набора данных с $n$ переменными — это квадратная матрица размером $n \times n$, где каждый элемент $r_{ij}$ представляет собой коэффициент корреляции Пирсона между переменными $X_i$ и $X_j$.

Обозначим набор переменных как $X_1, X_2, ..., X_n$. Корреляционная матрица $\mathbf{R}$ выглядит следующим образом:

$$
\mathbf{R} = \begin{pmatrix}
1 & r_{12} & r_{13} & \cdots & r_{1n} \\
r_{21} & 1 & r_{23} & \cdots & r_{2n} \\
r_{31} & r_{32} & 1 & \cdots & r_{3n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
r_{n1} & r_{n2} & r_{n3} & \cdots & 1
\end{pmatrix}
$$

где каждый элемент $r_{ij}$ — это коэффициент корреляции Пирсона между переменными $X_i$ и $X_j$:

$$
r_{ij} = \frac{\text{Cov}(X_i, X_j)}{\sigma_{X_i} \sigma_{X_j}}
$$

### Свойства корреляционной матрицы

1. **Диагональные элементы**: Все диагональные элементы корреляционной матрицы равны 1, поскольку корреляция любой переменной с самой собой равна 1.
2. **Симметрия**: Корреляционная матрица симметрична, т.е. $r_{ij} = r_{ji}$.
3. **Диапазон значений**: Все элементы корреляционной матрицы находятся в диапазоне от -1 до 1.

### Пример корреляционной матрицы

Рассмотрим пример набора данных с тремя переменными: $X_1$, $X_2$ и $X_3$. Допустим, что у нас есть следующие коэффициенты корреляции между этими переменными:

- $r_{12} = 0.8$ (корреляция между $X_1$ и $X_2$)
- $r_{13} = -0.2$ (корреляция между $X_1$ и $X_3$)
- $r_{23} = 0.5$ (корреляция между $X_2$ и $X_3$)

Корреляционная матрица для этого набора данных будет выглядеть так:

$$
\mathbf{R} = \begin{pmatrix}
1 & 0.8 & -0.2 \\
0.8 & 1 & 0.5 \\
-0.2 & 0.5 & 1
\end{pmatrix}
$$



### Использование корреляционной матрицы

1. **Анализ взаимосвязей**: Корреляционная матрица позволяет быстро оценить силу и направление линейных связей между всеми парами переменных в наборе данных.
2. **Выбор признаков**: В машинном обучении корреляционная матрица помогает выявить признаки, которые сильно коррелируют друг с другом. Сильная корреляция может указывать на избыточность, и один из таких признаков может быть исключен.
3. **Визуализация**: Корреляционная матрица часто визуализируется в виде тепловой карты, где цветовые градиенты отображают значения коэффициентов корреляции, что делает интерпретацию более наглядной.

### Вычисление корреляционной матрицы в Pandas

Загрузим датасет, чтобы проанализировать корреляцию между характеристиками

```python
import pandas as pd
from sklearn.datasets import load_wine

# Загрузка датасета Wine
wine = load_wine()

# Преобразование данных в DataFrame
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)

wine_df
```

Чтобы вычислить корреляционную матрицу в Pandas, нужно использовать функцию `corr`. Мы также передаём аргумент `numeric_only=True`, так как мы хотим посчитать корреляцию между всеми численными характеристиками.

```python
corr_matrix = wine_df.corr()
corr_matrix
```

Зачастую в виде таблицы изучать корреляцию неудобно, поэтому можно построить тепловую карту с помощью Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Визуализация корреляционной матрицы
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Матрица Корреляции')
plt.show()
```
