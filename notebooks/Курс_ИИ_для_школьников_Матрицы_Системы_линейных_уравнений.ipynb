## Матрицы

### Линейная комбинация в матричном виде

Возьмем три вектора

$$ \mathbf{u} = \begin{pmatrix} 1 \\ -1 \\ 0 \end{pmatrix}; \mathbf{v} = \begin{pmatrix} 0 \\ 1 \\ -1 \end{pmatrix}; \mathbf{w} = \begin{pmatrix} 1 \\ -1 \\ 0 \end{pmatrix} $$

Из них мы можем составить вот такую линейную комбинацию


$$ c \mathbf{u} + d \mathbf{v} + e \mathbf{w} = c \begin{pmatrix} 1 \\ -1 \\ 0 \end{pmatrix} + d \begin{pmatrix} 0 \\ 1 \\ -1 \end{pmatrix} + e \begin{pmatrix} 1 \\ -1 \\ 0 \end{pmatrix} = \begin{pmatrix} c \\ d-c \\ e - d \end{pmatrix} $$

Эту линейную комбинацию мы можем записать используя *матрицу*

$$
\begin{bmatrix} 1 & 0 & 1 \\ -1 & 1 & -1 \\ 0 & -1 & 0 \end{bmatrix}
\begin{bmatrix} c \\ d \\ e \end{bmatrix} =
\begin{bmatrix} c \\ d - c \\ e - d \end{bmatrix}
$$

Мы можем назвать матрицу буквой $A$:

$$ A = \begin{bmatrix} 1 & 0 & 1 \\ -1 & 1 & -1 \\ 0 & -1 & 0 \end{bmatrix} $$

Скаляры $c$, $d$, и $e$ мы можем переименовать в $x_1$, $x_2$, $x_3$ и вектор, состоящих из этих скаляров назвать вектором $\mathbf{x}$:

$$ \mathbf{x} = \begin{bmatrix} c \\ d \\ e \end{bmatrix} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix} $$

Вектор $ (c, d-c, e-d) $ мы можем назвать вектором $\mathbf{b}$ c компонентами $(b_1, b_2, b_3)$:

$$ \mathbf{b} = \begin{bmatrix} c \\ d - c \\ e - d \end{bmatrix} = \begin{bmatrix} x_1 \\ x_2 - x_1 \\ x_3 - x_2 \end{bmatrix} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix} $$

Наконец, мы можем записать линейную комбинацию как уравнение:

$$
\begin{bmatrix} 1 & 0 & 1 \\ -1 & 1 & -1 \\ 0 & -1 & 0 \end{bmatrix}
\begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix} =
\begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}
$$

что эквивалентно

$$ A \mathbf{x} = \mathbf{b} $$

### Линейные уравнения

До этого момента векторы $\mathbf{x}$ и $\mathbf{b}$ были для нас неизвестным: мы не знаем их конкретных значений. Но зачастую в задачах линейной алгебры неизвестным получается только вектор $\mathbf{x}$, в то время как конкретные значения компонент вектора $\mathbf{b}$ нам известны.

Теперь представим, что вектор $\mathbf{b}$ на известен и нам нужно найти вектор $\mathbf{x}$, чтобы уравнение выполнялось.

Таким образом, задача формулируются в следующем виде:

    Какие конкретно значения x1, x2, x3 нужно подставить в уравнение (линейную комбинацию),
    чтобы получить конкретные значения b1, b2, b3 в векторе b?  

Это задача напоминает задачу решения систем линейных уравнений. Действительно, если мы знаем все значения $b_1$, $b_2$, $b_3$, то мы можем переписать уравнение в виде системы линейных уравнений и решить её:

$$
\begin{cases}
x_1 = b_1 \\
-x_1 + x_2 = b_2 \\
-x_2 + x_3 = b_3
\end{cases}
$$

Решением этой системы будет:

$$
\begin{cases}
x_1 = b_1 \\
x_2 = b_1 + b_2 \\
x_3 = b_1 + b_2 + b_3
\end{cases}
$$

### Обратная матрица

Итак, уравнение $Ax = b$ имеет такое решение

$$ \mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix} = \begin{bmatrix} b_1 \\ b_2 + b_2 \\ b_1 + b_2 + b_3 \end{bmatrix} $$

Это решение можно также представить, используя матрицу $A^{-1}$

$$
\begin{bmatrix} b_1 \\ b_2 + b_2 \\ b_1 + b_2 + b_3 \end{bmatrix} =
\begin{bmatrix} 1 & 0 & 0 \\ 1 & 1 & 0 \\ 1 & 1 & 1 \end{bmatrix} \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}
$$

$$ \mathbf{x} = A^{-1} \mathbf{b}  $$

По сути, матрица $A^{-1}$ даёт нам решение системы линейных уравнений $A \mathbf{x} = \mathbf{b}$. Эта матрица называется **обратной матрицей**.

Если мы можем решить систему линейных уравнений, то говорят, что матрица $A$ **обратима**. Это значит, что используя обратимую матрицу $A$ мы можем вычислить обратную матрицу и найти вектор $\mathbf{x}$.

### Умножение матрицы на скаляр

При умножении матрицы на скаляр каждый элемент матрицы умножается на одно и то же число (скаляр). В результате получается новая матрица того же размера, что и исходная.

**Пример умножения матрицы на скаляр**

Рассмотрим матрицу $A$ и скаляр $k = 3$:

$$
A = \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}
$$

Умножим матрицу $A$ на $k$:

$$
3 \cdot A = 3 \cdot \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}
=
\begin{pmatrix}
3 \cdot 1 & 3 \cdot 2 \\
3 \cdot 3 & 3 \cdot 4
\end{pmatrix}
=
\begin{pmatrix}
3 & 6 \\
9 & 12
\end{pmatrix}
$$

### Сложение матриц

Матрицы можно также складывать друг с другом. Это операция, которая выполняется между двумя матрицами одинакового размера и заключается в сложении соответствующих элементов этих матриц. В результате получается новая матрица того же размера.

Рассмотрим пример с конкретными числами:

$$
A = \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix},
\quad
B = \begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix}
$$

Сумма $C = A + B$ вычисляется следующим образом:

$$
C = \begin{pmatrix}
1+5 & 2+6 \\
3+7 & 4+8
\end{pmatrix}
=
\begin{pmatrix}
6 & 8 \\
10 & 12
\end{pmatrix}
$$

### Правила и свойства матричных операций

**Форма (размер матрицы)**

Если матрица имеет $m$ строк и $n$ столбцов, то такая матрица имеет форму:
$$m \times n$$

**Обозначение элементов матрицы**

Число в матрице, находящееся в строке $i$ и колонке $j$ обозначаются как $a_{ij}$ или $A(i, j)$.

**Сложение матриц**

1. Две матрицы можно сложить друг с другом, если они имеют один и тот же размер

2. Сложение матриц обладает коммутативностью
$$ A + B = B + A $$

3. Сложение матриц обладает дистрибутивностью
$$ c(A + B) = cA + cB $$

3. Сложение матриц обладает ассоциативностью
$$ (A + B) + С = A + (B + С) $$

**Умножение матриц**

1. Чтобы умножить матрицу $A$ на матрицу $B$, количество колонок в матрице $A$ должно совпадать с количеством строк в матрице $B$

2. Если матрица $A$ имеет форму $m \times n$ и матрица $B$ имеет форму $n \times p$, то матрица, полученная в результате умножения, будет иметь форму

$$ (m \times n)(n \times p) = (m \times p) $$

3. Умножение матриц *не обладает коммутативностью*
$$ AB \neq BA $$

4. Умножение матриц обладает дистрибутивностью слева
$$ C(A + B) = CA + CB $$

5. Умножение матриц обладает дистрибутивностью справа
$$ (A + B)С = AC + BC $$

6. Умножение матриц обладает ассоциативностью
$$ (AB)С = A(BC) $$

### Единичная матрица

**Единичная матрица** (иногда называемая **матричной единицей** или **матрицей идентичности**) — это квадратная матрица, у которой на главной диагонали стоят единицы, а все остальные элементы равны нулю. Единичная матрица обозначается как $I_n$, где $n$ — размер матрицы (количество строк или столбцов).

**Пример единичной матрицы**

Для размера $n = 2$, единичная матрица выглядит так:

$$
I_2 = \begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
$$

Для $n = 3$:

$$
I_3 = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
$$


**Свойства единичной матрицы**

1. **Матричное умножение**: Единичная матрица играет ту же роль в матричном умножении, что и число 1 в обычном умножении чисел. Если умножить любую матрицу $A$ на единичную матрицу, результат будет равен самой матрице $A$:
   $$
   I_n \cdot A = A \cdot I_n = A
   $$
   Это свойство действует для любой квадратной матрицы $A$ размером $n \times n$.

2. **Обратная матрица**: Единичная матрица является обратной для самой себя. Если матрицу $A$ умножить на её обратную матрицу $A^{-1}$, результат будет равен единичной матрице:
   $$
   A \cdot A^{-1} = A^{-1} \cdot A = I_n
   $$

### Ступенчатый (треугольный) вид матрицы

Ступенчатый (или треугольный) вид матрицы — это форма матрицы, в которой элементы ниже главной диагонали равны нулю.

**Виды треугольных матриц**

1. Верхнетреугольная матрица
   
   $$
   A = \begin{pmatrix}
   a_{11} & a_{12} & a_{13} \\
   0 & a_{22} & a_{23} \\
   0 & 0 & a_{33} \\
   \end{pmatrix}
   $$

2. Нижнетреугольная матрица

   $$
   B = \begin{pmatrix}
   b_{11} & 0 & 0 \\
   b_{21} & b_{22} & 0 \\
   b_{31} & b_{32} & b_{33} \\
   \end{pmatrix}
   $$

**Ступенчатый вид**

Матрица находится в ступенчатом виде, если выполнены следующие условия:

1. Первый ненулевой элемент каждой строки, называемый *ведущим элементом*, расположен правее, чем в предыдущей строке.
2. Все строки, состоящие из нулей, находятся ниже строк с ненулевыми элементами.

Пример ступенчатого вида матрицы:

$$
C = \begin{pmatrix}
1 & 2 & 3 & 4 \\
0 & 5 & 6 & 7 \\
0 & 0 & 8 & 9 \\
\end{pmatrix}
$$

### Система из двух линейных уравнений

Рассмотрим такую систему уравнений

$$
\begin{cases}
x - 2y = 1 \\
3x + 2y = 11
\end{cases}
$$

Визуализируем её на плоскостн $xOy$. Первое уравнение $x - 2y = 1$ создает одну линию на плоскости, а уравнение $2x + 2y = 11$ создает другую линию на плоскости. Точка $x = 3, y = 1$ лежит на обоих линиях. Эта точка пересечения линий и решает систему.

```python
import numpy as np
import matplotlib.pyplot as plt

# Определим диапазон значений x
x_values = np.linspace(-10, 10, 400)

# Определим функции прямых, полученных из уравнений системы
y1 = (x_values - 1) / 2  # Первая прямая: x - 2y = 1 -> y = (x - 1) / 2
y2 = (11 - 3*x_values) / 2  # Вторая прямая: 2x + 2y = 11 -> y = (11 - 2x) / 2

# Найдем точку пересечения прямых, решив систему уравнений
intersection_x = 3
intersection_y = 1

# Построим графики
plt.figure(figsize=(8, 8))
plt.plot(x_values, y1, label=r'$x - 2y = 1$')
plt.plot(x_values, y2, label=r'$3x + 2y = 11$')

# Подписи прямых
plt.annotate(r'$x - 2y = 1$', xy=(5, 2), xytext=(6, 4))
plt.annotate(r'$2x + 2y = 11$', xy=(2, 4.5), xytext=(-5, 7))

# Подпись и выделение точки пересечения
plt.scatter(intersection_x, intersection_y, color='red')  # Выделим точку пересечения
plt.annotate(
    r'$x = 3, y = 1$',
    xy=(intersection_x, intersection_y),
    xytext=(10, 0),
    textcoords='offset points',
    ha='left',
    va='center'
)

# Добавим оси координат
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Настройки графика
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.legend()
plt.grid(True)
plt.title('Графики прямых, заданных системой уравнений')
plt.xlabel('x')
plt.ylabel('y')

# Отобразим график
plt.show()
```

Эту систему можно представить как матричное уравнение:

$$ A \mathbf{x} = \mathbf{b} $$

$$
\begin{bmatrix} 1 & -2 \\ 3 & 2 \end{bmatrix}
\begin{bmatrix} x \\ y \end{bmatrix} =
\begin{bmatrix} 1 \\ 11 \end{bmatrix}
$$

Обратив матрицу $A$ мы можем найти решение системы.

### Система из трех линейных уравнений

Рассмотрим другую систему уравнений

$$
\begin{cases}
x + 2y + 3z = 6 \\
2x + 5y + 2z = 4 \\
6x - 3y + z = 2 \\
\end{cases}
$$

Визуализируем эту систему в 3D пространстве. Каждое из уравнений системы будут создавать плоскость в пространстве. Аналогично предыдущей системе уравнений, решением будет пересечение.

```python
import numpy as np
import plotly.graph_objs as go

# Определим диапазон значений для x и y
x_values = np.linspace(-10, 10, 50)
y_values = np.linspace(-10, 10, 50)

# Создадим сетку значений для x и y
X, Y = np.meshgrid(x_values, y_values)

# Выразим z из каждого уравнения
Z1 = (6 - X - 2*Y) / 3  # x + 2y + 3z = 6 -> z = (6 - x - 2y) / 3
Z2 = (4 - 2*X - 5*Y) / 2  # 2x + 5y + 2z = 4 -> z = (4 - 2x - 5y) / 2
Z3 = (2 - 6*X + 3*Y)  # 6x - 3y + z = 2 -> z = 2 - 6x + 3y

# Определим плоскости для графиков
plane1 = go.Surface(x=X, y=Y, z=Z1, colorscale='Viridis', opacity=0.6, name='x + 2y + 3z = 6')
plane2 = go.Surface(x=X, y=Y, z=Z2, colorscale='Cividis', opacity=0.6, name='2x + 5y + 2z = 4')
plane3 = go.Surface(x=X, y=Y, z=Z3, colorscale='Plasma', opacity=0.6, name='6x - 3y + z = 2')

# Определим точку пересечения плоскостей
intersection_point = go.Scatter3d(x=[0], y=[0], z=[2],
                                  mode='markers',
                                  marker=dict(size=10, color='red'),
                                  name='Решение')

# Определим макет и добавим плоскости на график
layout = go.Layout(title='Визуализация системы уравнений в 3D',
                   scene=dict(xaxis_title='x', yaxis_title='y', zaxis_title='z'))

fig = go.Figure(data=[plane1, plane2, plane3, intersection_point], layout=layout)

# Отобразим график
fig.show()
```

Эту систему также можно представить как матричное уравнение:

$$ A \mathbf{x} = \mathbf{b} $$

$$
\begin{bmatrix} 1 & 2 & 3 \\ 2 & 5 & 2 \\ 6 & -3 & 1 \end{bmatrix}
\begin{bmatrix} x \\ y \\ z \end{bmatrix} =
\begin{bmatrix} 6 \\ 4 \\ 2 \end{bmatrix}
$$

Обратив матрицу $A$ мы можем найти решение системы.

### Решение линейных уравнений с помощью исключения

Мы можем манипулировать уравнениями, чтобы решить систему таких уравнений. Один из таких методов решения - метод исключения.

Пусть у нас есть такая система уравнений

$$
\begin{cases}
x - 2y = 1 \\
3x + 2y = 11
\end{cases}
$$

Использовав метод исключения, мы можем получить вот такую систему, которая будет равносильна предыдущей системе

$$
\begin{cases}
x - 2y = 1 \\
8y = 8
\end{cases}
$$

В системе поменялось только одно уравнение: второе. Оно специально приведено именно к такому уравнению, когда в уравнении есть только одно неизвестное, чтобы его можно было сразу найти (в этом случае, $y$ будет равен 1).

Подставив $y = 1$ в первое уравнение

$$ x - 2 \cdot 1 = 1 $$

мы можем найти и $x$, который будет равен 3.

**Как мы выполнили метод исключения, чтобы получить вторую систему?**

1. Мы умножили первое уравнение на 3 и получили такое уравнение

$$2x - 6y = 3$$

2. Мы вычли полученное уравнение из второго уравнения в системе

$$(3x - 2y) - (2x - 6y) = 11 - 3$$

$$8y = 8$$

Таким образом, мы **исключили** $x$ из второго уравнения, тем самым оставив только одно неизвестное.

### Расширенная матрица

Для решения систем линейных уравнений используют так называемую **расширенную матрицу**. Она включает в себя как коэффициенты переменных, так и свободные члены (правая часть уравнений). Расширенная матрица строится путём добавления столбца свободных членов к матрице коэффициентов системы.

Рассмотрим систему двух уравнений с двумя переменными:

$$
\begin{cases}
2x_1 + 3x_2 = 5 \\
4x_1 + 6x_2 = 8
\end{cases}
$$

Матрица коэффициентов будет:

$$
A = \begin{pmatrix}
2 & 3 \\
4 & 6
\end{pmatrix}
$$

Вектор свободных членов:

$$
\mathbf{b} = \begin{pmatrix}
5 \\
8
\end{pmatrix}
$$

Расширенная матрица для этой системы:

$$
\left[\begin{array}{cc|c}
2 & 3 & 5 \\
4 & 6 & 8
\end{array}\right]
$$



### Решение системы линейных уравнений с помощью метода Жордана-Гаусса

Метод Жордана-Гаусса заключается в преобразовании матрицы системы к ступенчатому виду или к единичной матрице с помощью элементарных преобразований строк.

**Пример решения системы уравнений:**

Рассмотрим простую систему линейных уравнений:
$$
\begin{cases}
2x_1 + x_2 = 5 \\
4x_1 + 3x_2 = 11
\end{cases}
$$

Записываем расширенную матрицу:
$$
\left[\begin{array}{cc|c}
2 & 1 & 5 \\
4 & 3 & 11
\end{array}\right]
$$

Преобразуем первую строку, чтобы получить единицу в первом столбце (делим первую строку на 2):
$$
\left[\begin{array}{cc|c}
1 & \frac{1}{2} & \frac{5}{2} \\
4 & 3 & 11
\end{array}\right]
$$

Вычитаем из второй строки первую, умноженную на 4, чтобы получить 0 в первом элементе второй строки:
$$
\left[\begin{array}{cc|c}
1 & \frac{1}{2} & \frac{5}{2} \\
0 & 1 & 1
\end{array}\right]
$$

Теперь, умножаем вторую строку на $\frac{1}{2}$ и вычитаем ее из первой строки:
$$
\left[\begin{array}{cc|c}
1 & 0 & 2 \\
0 & 1 & 1
\end{array}\right]
$$

Решение системы: $x_1 = 2$, $x_2 = 1$.

Таким образом, метод Жордана-Гаусса позволяет найти решение системы линейных уравнений, приводя матрицу системы к единичной матрице.
