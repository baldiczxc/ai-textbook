Знания математического анализа играют ключевую роль в машинном обучении, поскольку они помогают лучше понимать и эффективно использовать алгоритмы и методы оптимизации. Вот несколько примеров, где математический анализ особенно полезен:

1. **Градиентный спуск:** Для понимания того, как алгоритмы машинного обучения обучаются, необходимо понимать что такое градиент (или производная), который является инструментом математического анализа.

## Теория пределов

### Последовательность

Последовательность — это упорядоченный набор чисел, расположенных в определённом порядке. Последовательности часто обозначаются как $a_1, a_2, a_3, \dots$, где каждый элемент $a_n$ соответствует своему номеру $n$.

Формально, последовательность — это функция, которая сопоставляет каждому натуральному числу $n$ некоторое значение $a_n$. В математической записи это выглядит так:

$$a_n = f(n)$$

где $f(n)$ — это правило, по которому определяется значение $n$-го элемента последовательности.



**Примеры последовательностей**

* *Арифметическая последовательность*
   
   В этой последовательности разность между любыми двумя последовательными членами одинакова. Например:
   $$a_n = 1 + 2(n - 1)$$
   Это последовательность $1, 3, 5, 7, \dots$

* *Геометрическая последовательность*
   
   Здесь каждый следующий элемент получается умножением предыдущего на постоянный коэффициент. Например:
   $$a_n = 2 \cdot 3^{n-1}$$
   Это последовательность $2, 6, 18, 54, \dots$

### Предел последовательности

Предел последовательности описывает поведение последовательности по мере того, как номер её члена стремится к бесконечности.

Представьте последовательность чисел
$$a_1, a_2, a_3, \dots$$

Мы можем спросить: «К какому числу приближаются элементы последовательности по мере увеличения номера $n$?» Если существует такое число $L$, к которому элементы последовательности становятся всё ближе и ближе, то это число называется пределом последовательности.

#### Формальное определение

Последовательность $a_n$ имеет предел $L$, если для любого сколь угодно малого положительного числа $\epsilon$ можно найти такое натуральное число $N$, что для всех $n > N$ выполняется неравенство:

$$|a_n - L| < \epsilon$$

Иными словами, начиная с некоторого номера $N$, все элементы последовательности находятся в пределах $L \pm \epsilon$.

#### Пример



Рассмотрим последовательность $a_n = \frac{1}{n}$. По мере увеличения $n$, значения этой последовательности становятся всё ближе и ближе к нулю.

```python
import matplotlib.pyplot as plt

# Задаем количество элементов последовательности
n_values = range(1, 51)

# Рассчитываем значения последовательности 1/n
sequence = [1/n for n in n_values]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence, marker='o', linestyle='-', color='b', label=r'$a_n = \frac{1}{n}$')

# Настройка осей и заголовка
plt.title('Визуализация последовательности $a_n = 1/n$')
plt.xlabel('n')
plt.ylabel(r'$a_n$')
plt.ylim(0, 1)
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```

Мы можем сказать, что предел этой последовательности равен 0:

$$\lim_{n \to \infty} \frac{1}{n} = 0$$

Здесь, если мы выберем любое маленькое $\epsilon > 0$, то можно найти такое число $N$, что для всех $n > N$ выполняется $|\frac{1}{n} - 0| < \epsilon$.


#### Сходящиеся и расходящиеся последовательности

1. **Сходящиеся последовательности** — это последовательности, которые имеют предел. Например, в случае последовательности $a_n = \frac{1}{n}$ её предел равен 0.

2. **Расходящиеся последовательности** — это последовательности, которые не имеют конечного предела. Например, последовательность $a_n = n$ стремится к бесконечности, и поэтому у неё нет конечного предела.


#### Бесконечный предел

Последовательность может также стремиться к бесконечности. Например, если $a_n = n^2$, то:

$$\lim_{n \to \infty} n^2 = \infty$$

Это означает, что по мере увеличения $n$ элементы последовательности растут без ограничений.

### Что такое функция?

Функция — это взаимосвязь между величинами, то есть зависимость одной переменной величины от другой.

$$ y = f(x) $$

Под функцией понимают правило, формулу, уравнение, которое описывает зависимость одной переменной от другой (например, у от х). Если изучить функцию, мы поймем:

* как изменится одна переменная, если другая увеличится;

* что произойдет с аргументом, если мы уменьшим функцию;

* что будет, если мы отобразим эту зависимость графически.

Важное напоминание: функция — это зависимая переменная величина (чаще у), аргумент — независимая переменная (чаще х).

### Предел функции

Предел функции описывает поведение функции вблизи определенной точки. Оно используется для понимания того, к какому значению стремится функция, когда её аргумент приближается к некоторому числу.

Представьте функцию $f(x)$, и пусть $x$ приближается к какому-то числу $c$. Если значения функции $f(x)$ приближаются к определённому числу $L$, то говорят, что предел функции $f(x)$ при $x$ стремящемся к $c$ равен $L$.

Записывается это так:

$$\lim_{x \to c} f(x) = L$$



Предел функции используется для анализа её поведения в окрестности точек разрыва, при рассмотрении асимптот, а также в определении производных и интегралов. Понимание предела функции позволяет лучше понять, как ведет себя функция в тех или иных ситуациях, особенно в сложных случаях.

#### Формальное определение



Формально предел функции $f(x)$ при $x \to c$ равен $L$, если для любого сколь угодно малого положительного числа $\epsilon$ существует такое положительное число $\delta$, что для всех $x$, удовлетворяющих условию $0 < |x - c| < \delta$, выполняется:

$$|f(x) - L| < \epsilon$$

Это означает, что когда $x$ находится достаточно близко к $c$, значение функции $f(x)$ будет достаточно близко к $L$.

#### Пример





Рассмотрим функцию $f(x) = 2x$. Найдём предел этой функции при $x$ стремящемся к 3:

$$\lim_{x \to 3} 2x = 2 \times 3 = 6$$

Здесь, по мере того как $x$ приближается к 3, значение функции $f(x)$ приближается к 6.

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x):
    return 2 * x

# Диапазон значений x
x_values = np.linspace(2, 4, 400)
y_values = f(x_values)

# Вычисляем значение функции в точке x=3
x_target = 3
y_target = f(x_target)

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$f(x) = 2x$', color='b')
plt.scatter([x_target], [y_target], color='r')  # Точка (3, 6)

# Добавляем линии, показывающие x=3 и f(3)=6
plt.axvline(x=x_target, ymax=0.5, color='g', linestyle='--', label=r'$x = 3$')
plt.axhline(y=y_target, xmax=0.5, color='orange', linestyle='--', label=r'$f(3) = 6$')

# Настройка осей и заголовка
plt.title('Предел функции $f(x) = 2x$ при $x$ стремящемся к 3')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```

#### Односторонние пределы


Иногда полезно рассматривать пределы функции с разных сторон:

- **Левосторонний предел** — когда $x$ стремится к $c$ слева (обозначается $x \to c^-$).
  
- **Правосторонний предел** — когда $x$ стремится к $c$ справа (обозначается $x \to c^+$).

Если левосторонний и правосторонний пределы равны, то говорят, что существует общий предел функции в точке $c$.

### Бесконечные пределы



Если функция растёт без ограничения по мере того, как $x$ приближается к $c$, то говорят, что предел равен бесконечности:

$$\lim_{x \to c} f(x) = \infty$$

Аналогично, если функция стремится к минус бесконечности, то предел равен $-\infty$.

Рассмотрим функцию $f(x) = \frac{1}{x}$ и найдём её предел при $x$ стремящемся к нулю:

- При $x \to 0^+$ (с положительных значений), $f(x)$ стремится к $+\infty$.
- При $x \to 0^-$ (с отрицательных значений), $f(x)$ стремится к $-\infty$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x):
    return 1 / x

# Диапазон значений x с исключением нуля (чтобы избежать деления на 0)
neg_x_values = np.linspace(-2, -0.1, 400).tolist()
pos_x_values = np.linspace(0.1, 2, 400).tolist()
neg_y_values = f(np.array(neg_x_values))
pos_y_values = f(np.array(pos_x_values))

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(neg_x_values, neg_y_values, label=r'$f(x) = \frac{1}{x}$', color='b')
plt.plot(pos_x_values, pos_y_values, color='b')

# Добавляем линии, показывающие x=0
plt.axvline(x=0, color='black', linestyle='--', label=r'$x = 0$')
plt.axhline(y=0, color='black', linestyle='--', label=r'$x = 0$')

# Настройка осей и заголовка
plt.ylim(-10, 10)
plt.title(r'Предел функции $f(x) = \frac{1}{x}$ при $x$ стремящемся к 0')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.legend()

# Показать график
plt.show()
```

### Вычисление пределов функции

#### Прямое подставление

Самый простой способ вычисления предела функции — это прямое подставление значения $x = c$ в функцию $f(x)$. Если функция непрерывна в точке $c$, то её предел при $x \to c$ равен значению функции в этой точке:

$$\lim_{x \to c} f(x) = f(c)$$

**Пример**: Для функции $f(x) = 2x + 3$:

$$\lim_{x \to 2} (2x + 3) = 2 \times 2 + 3 = 7$$

#### Разложение на множители

Если прямое подставление приводит к неопределённости типа $0/0$, можно попробовать разложить числитель и знаменатель на множители и упростить выражение. Часто это помогает устранить неопределённость и вычислить предел.

**Пример**: Найдём предел функции $f(x) = \frac{x^2 - 4}{x - 2}$ при $x \to 2$:

Разложим числитель на множители:

$$f(x) = \frac{(x - 2)(x + 2)}{x - 2}$$

Сократим $(x - 2)$:

$$f(x) = x + 2$$

Теперь можем подставить $x = 2$:

$$\lim_{x \to 2} (x + 2) = 2 + 2 = 4$$

#### Использование предельных теорем

Для вычисления пределов можно использовать различные теоремы:

- **Теорема о сумме**: Предел суммы двух функций равен сумме их пределов.

$$\lim_{x \to c} [f(x) + g(x)] = \lim_{x \to c} f(x) + \lim_{x \to c} g(x)$$

- **Теорема о произведении**: Предел произведения двух функций равен произведению их пределов.

$$\lim_{x \to c} [f(x) \cdot g(x)] = \lim_{x \to c} f(x) \cdot \lim_{x \to c} g(x)$$

- **Теорема о частном**: Предел частного равен частному пределов, если предел знаменателя не равен нулю.

$$\lim_{x \to c} \frac{f(x)}{g(x)} = \frac{\lim_{x \to c} f(x)}{\lim_{x \to c} g(x)}, \quad \text{если} \ \lim_{x \to c} g(x) \neq 0$$



#### Пределы на бесконечности

Для вычисления пределов при $x \to \infty$ или $x \to -\infty$, важно понимать асимптотическое поведение функции. Можно анализировать, к какому значению стремится функция, если её аргумент становится очень большим по модулю.

**Пример**: Найдём предел $\lim_{x \to \infty} \frac{1}{x}$.

Когда $x$ становится очень большим, $1/x$ стремится к нулю:

$$\lim_{x \to \infty} \frac{1}{x} = 0$$

Таким образом, существует множество методов для вычисления пределов функции, и выбор метода зависит от конкретной ситуации и вида функции.

#### Продвинутые приемы

В некоторых случаях использование тех приемов, которые перечислены выше, недостаточно для вычисления предела. В таких случаях могут использовать более продвинутые приемы, такие как *замена переменной* и *правило Лопиталя*.
