## Дифференциальное исчисление

### Производная


Производная — это одно из центральных понятий математического анализа, которое позволяет понять, как быстро изменяется значение функции при изменении её аргумента.

#### Формальное определение производной





Пусть $f(x)$ — функция, и $x$ — переменная. Производная функции $f(x)$ в точке $x = a$ определяется как предел отношения изменения значения функции к изменению аргумента при стремлении этого изменения аргумента к нулю:

$$f'(a) = \lim_{\Delta x \to 0} \frac{f(a + \Delta x) - f(a)}{\Delta x}$$

Здесь:

- $f'(a)$ — производная функции $f(x)$ в точке $x = a$.
- $\Delta x$ — маленькое изменение аргумента $x$.
- $f(a + \Delta x) - f(a)$ — изменение значения функции.

Производная $f'(x)$ ещё также обозначается с использованием *дифференциалов*:

$$f'(a) = \frac{d}{dx}f(a) = \frac{df(a)}{dx}$$

Производная $f'(x)$ в точке $x = a$ показывает, как изменяется функция при малом изменении $x$ на величину $\Delta x$.

#### Геометрический смысл производной

Геометрически производная — это угловой коэффициент касательной к графику функции в данной точке. Если график функции в точке $x = a$ имеет положительный наклон, то производная положительна, если отрицательный наклон — производная отрицательна.

#### Производная для функции $f(x) = x^2$

Рассмотрим функцию $f(x) = x^2$. Найдём её производную:

$$f'(x) = \lim_{\Delta x \to 0} \frac{(x + \Delta x)^2 - x^2}{\Delta x}$$

Раскроем скобки в числителе:

$$f'(x) = \lim_{\Delta x \to 0} \frac{x^2 + 2x\Delta x + (\Delta x)^2 - x^2}{\Delta x}$$

Упростим выражение:

$$f'(x) = \lim_{\Delta x \to 0} \frac{2x\Delta x + (\Delta x)^2}{\Delta x}$$

Сократим $\Delta x$:

$$f'(x) = \lim_{\Delta x \to 0} (2x + \Delta x)$$

Когда $\Delta x$ стремится к нулю, остаётся:

$$f'(x) = 2x$$

Таким образом, производная функции $f(x) = x^2$ равна $2x$. Это означает, что в любой точке $x$ скорость изменения функции $f(x)$ равна $2x$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию и её производную
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# Диапазон значений x
x_values = np.linspace(-10, 10, 400)
y_values = f(x_values)
y_prime_values = f_prime(x_values)

# Создание фигуры и двух подграфиков
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# График функции f(x) = x^2 на первом подграфике
ax1.plot(x_values, y_values, label=r'$f(x) = x^2$', color='b')
ax1.set_title('График функции $f(x) = x^2$')
ax1.axhline(0, color='black',linewidth=0.5)
ax1.axvline(0, color='black',linewidth=0.5)
ax1.grid(True)
ax1.legend()

# График производной f'(x) = 2x на втором подграфике
ax2.plot(x_values, y_prime_values, label=r"$f'(x) = 2x$", color='r')
ax2.set_title('График производной $f\'(x) = 2x$')
ax2.axhline(0, color='black',linewidth=0.5)
ax2.axvline(0, color='black',linewidth=0.5)
ax2.grid(True)
ax2.legend()

# Отображение графиков
plt.tight_layout()
plt.show()
```

#### Правила дифференцирования

---

**Линейное свойство (правило дифференцирования суммы и множителей)**

---

Линейное свойство дифференцирования гласит, что производная суммы функций равна сумме их производных, а производная функции, умноженной на константу, равна произведению константы и производной этой функции.

Если $f(x)$ и $g(x)$ — дифференцируемые функции, и $c$ — константа, то:

$$
\frac{d}{dx}[f(x) + g(x)] = f'(x) + g'(x)
$$

$$
\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)
$$

Пример: Пусть $f(x) = 2x^2 + 3x$. Производная будет:

$$
f'(x) = \frac{d}{dx}[2x^2] + \frac{d}{dx}[3x] = 2 \cdot 2x + 3 \cdot 1 = 4x + 3
$$

---

**Производная произведения**

---

Производная произведения двух функций $f(x)$ и $g(x)$ вычисляется как:

$$
\frac{d}{dx}[f(x) \cdot g(x)] = f'(x) \cdot g(x) + f(x) \cdot g'(x)
$$

Здесь важно заметить, что сначала дифференцируется первая функция с умножением на вторую, а затем дифференцируется вторая функция с умножением на первую.

Пример: Пусть $f(x) = x^2$ и $g(x) = \sin(x)$. Тогда производная произведения этих функций будет:

$$
\frac{d}{dx}[x^2 \cdot \sin(x)] = 2x \cdot \sin(x) + x^2 \cdot \cos(x)
$$



---

**Производная частного**

---

Производная частного двух функций $f(x)$ и $g(x)$ вычисляется по следующему правилу:

$$
\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{[g(x)]^2}
$$

В числителе стоит разность произведений производной числителя на знаменатель и числителя на производную знаменателя. Знаменатель возводится в квадрат.

Пример: Пусть $f(x) = x^2$ и $g(x) = x + 1$. Тогда производная частного этих функций будет:

$$
\frac{d}{dx}\left[\frac{x^2}{x+1}\right] = \frac{(2x \cdot (x + 1)) - (x^2 \cdot 1)}{(x + 1)^2} = \frac{2x^2 + 2x - x^2}{(x + 1)^2} = \frac{x^2 + 2x}{(x + 1)^2}
$$



---

**Производная сложной функции**

---

Производная сложной функции (правило цепочки) используется, когда функция представлена как композиция двух или более функций. Пусть $y = f(g(x))$, тогда:

$$
\frac{dy}{dx} = f'(g) \cdot g'(x)
$$

Здесь дифференцируется внешняя функция по внутренней, а затем внутреннюю функцию по аргументу.

Пример: Пусть $y = \sin(x^2)$. Тогда, согласно правилу цепочки:

$$
\frac{dy}{dx} = \cos(x^2) \cdot 2x
$$

#### Пример вычисления производной

Найдите производную функции $f(x) = 3x^4 - 5x^3 + 2x^2 - 7x + 4$.

1. **Применим правило дифференцирования:**

   Производная суммы функций равна сумме производных, а производная каждого члена многочлена находится по правилу из *таблицы прозводных*:

   $$\frac{d}{dx}(ax^n) = n \cdot ax^{n-1}$$

2. **Найдем производные для каждого члена функции:**

   - Производная от $3x^4$:

     $$
     \frac{d}{dx}(3x^4) = 4 \cdot 3x^{4-1} = 12x^3
     $$

   - Производная от $-5x^3$:

     $$
     \frac{d}{dx}(-5x^3) = 3 \cdot (-5)x^{3-1} = -15x^2
     $$

   - Производная от $2x^2$:

     $$
     \frac{d}{dx}(2x^2) = 2 \cdot 2x^{2-1} = 4x
     $$

   - Производная от $-7x$:

     $$
     \frac{d}{dx}(-7x) = -7
     $$

   - Производная от константы $4$:

     $$
     \frac{d}{dx}(4) = 0
     $$

3. **Сложим все найденные производные:**

   $$f'(x) = 12x^3 - 15x^2 + 4x - 7$$

Производная функции $f(x) = 3x^4 - 5x^3 + 2x^2 - 7x + 4$ равна $f'(x) = 12x^3 - 15x^2 + 4x - 7$.

### Исследование функций с помощью производных

#### Монотонность

**Определение монотонности**

Начнем с того, что такое монотонность функции:





- **Возрастающая функция**: Функция $f(x)$ называется возрастающей на интервале $I$, если для любых $x_1$ и $x_2$ из $I$ таких, что $x_1 < x_2$, выполняется неравенство $f(x_1) \leq f(x_2)$.



498px-Monotonicity_example1.png

- **Убывающая функция**: Функция $f(x)$ называется убывающей на интервале $I$, если для любых $x_1$ и $x_2$ из $I$ таких, что $x_1 < x_2$, выполняется неравенство $f(x_1) \geq f(x_2)$.

498px-Monotonicity_example2.png

**Связь между производной и монотонностью функции**

Теперь рассмотрим как производная связана с монотонностью функции:



- **Если производная положительна**: Если $f'(x) > 0$ для всех $x$ на интервале $I$, то функция $f(x)$ возрастает на этом интервале.



- **Если производная отрицательна**: Если $f'(x) < 0$ для всех $x$ на интервале $I$, то функция $f(x)$ убывает на этом интервале.



- **Если производная равна нулю**: Если $f'(x) = 0$ для всех $x$ на интервале $I$, то функция $f(x)$ постоянна на этом интервале.

#### Минимумы и максимумы

**Локальный максимум и минимум**

- **Локальный максимум**: Функция $f(x)$ имеет локальный максимум в точке $x_0$, если существует окрестность точки $x_0$ такая, что для всех $x$ из этой окрестности выполняется неравенство $f(x) \leq f(x_0)$. Это означает, что значение функции в точке $x_0$ больше, чем во всех ближайших точках.

- **Локальный минимум**: Функция $f(x)$ имеет локальный минимум в точке $x_0$, если существует окрестность точки $x_0$ такая, что для всех $x$ из этой окрестности выполняется неравенство $f(x) \geq f(x_0)$. Это означает, что значение функции в точке $x_0$ меньше, чем во всех ближайших точках.

Важно подчеркнуть, что локальный максимум и минимум называются **локальными**, так как это утверждение верно только для некоторой окрестности точки $x_0$, но не обязательно для всей области определения функции.


**Локальный экстремум**

Это общее название для локальных максимумов и минимумов. Таким образом, если функция имеет в точке $x_0$ либо локальный максимум, либо локальный минимум, то говорят, что в этой точке у функции локальный экстремум.



**Теорема Ферма**

> Если функция $f(x)$ имеет локальный экстремум в точке $x_0$ и при этом функция дифференцируема в этой точке, то $f'(x_0) = 0$.

Эта теорема является основным инструментом для нахождения кандидатов на экстремумы, так как она указывает на то, что в точках экстремума производная функции равна нулю.

**Недостаточность теоремы Ферма**

Теорема Ферма не является достаточным условием для того, чтобы точка $x_0$ была экстремумом. Она только говорит, что если экстремум есть, то в этой точке производная равна нулю. Однако обратное неверно: если $f'(x_0) = 0$, это не обязательно означает, что $x_0$ — точка экстремума. Например, точка перегиба также может иметь нулевую производную, но не быть экстремумом.

Примером такого случая может служить функция $x^3$

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем диапазон значений x
x = np.linspace(-2, 2, 400)

# Задаем функцию и её производную
y = x**3
y_prime = 3 * x**2

# Создаем фигуру с двумя подграфиками
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Первый подграфик - функция y = x^3
axs[0].plot(x, y, label='$y = x^3$', color='blue')
axs[0].axvline(x=0, color='black', linestyle='--')  # Вертикальная пунктирная линия в x = 0
axs[0].grid(True)
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('График функции $y = x^3$')
axs[0].legend()

# Второй подграфик - производная y' = 3x^2
axs[1].plot(x, y_prime, label="$y' = 3x^2$", color='red')
axs[1].axvline(x=0, color='black', linestyle='--')  # Вертикальная пунктирная линия в x = 0
axs[1].grid(True)
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('График производной $y\' = 3x^2$')
axs[1].legend()

# Уменьшаем отступы между подграфиками
plt.tight_layout()

# Показываем графики
plt.show()
```

**Стационарные точки**

Точка $x_0$ называется стационарной, если производная функции в этой точке равна нулю, то есть $f'(x_0) = 0$. Стационарные точки — это кандидаты на локальные экстремумы.

Однако не каждая стационарная точка является точкой экстремума, поэтому для окончательной классификации стационарных точек используются дополнительные методы, такие как исследование второй производной.

**Второе достаточное условие экстремума**

> Пусть функция $f(x)$ дважды дифференцируема в точке $x_0$ и $f'(x_0) = 0$. Если при этом $f''(x_0) < 0$, то $x_0$ - точка максимума, а если $f''(x_0)$, то точка минимума.

Теперь, следуя этому условию для экстремума, мы уже не можем сказать, что $x = 0$ это экстремум функции $f(x) = x^3$

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию и её вторую производную
def f(x):
    return x**3

def f_double_prime(x):
    return 6 * x

# Диапазон значений x
x = np.linspace(-2, 2, 400)

# Значения функции и второй производной
y = f(x)
y_double_prime = f_double_prime(x)

# Создаем фигуру с двумя подграфиками
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# График функции y = x^3
axs[0].plot(x, y, label='$f(x) = x^3$', color='blue')
axs[0].axvline(x=0, color='black', linestyle='--')  # Вертикальная пунктирная линия в x = 0
axs[0].grid(True)
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].set_title('График функции $f(x) = x^3$')
axs[0].legend()

# График второй производной y'' = 6x
axs[1].plot(x, y_double_prime, label="$f''(x) = 6x$", color='red')
axs[1].axvline(x=0, color='black', linestyle='--')  # Вертикальная пунктирная линия в x = 0
axs[1].grid(True)
axs[1].set_xlabel('x')
axs[1].set_ylabel("f''(x)")
axs[1].set_title('График второй производной $f\'\'(x) = 6x$')
axs[1].legend()

# Уменьшаем отступы между подграфиками
plt.tight_layout()

# Показываем графики
plt.show()
```

#### Пример поиска максимума

Найдите точку максимума функции $f(x) = -2x^2 + 4x + 1$ и определите значение функции в этой точке.

1. **Найдем производную функции:**

   Чтобы найти точку максимума, сначала вычислим производную функции $f(x)$:

   $$
   f'(x) = \frac{d}{dx}(-2x^2 + 4x + 1) = -4x + 4
   $$

2. **Найдем экстремумы:**

   Экстремумы — это значения $x$, при которых производная $f'(x)$ равна нулю:

   $$
   f'(x) = -4x + 4 = 0
   $$

   Решаем это уравнение:

   $$
   -4x + 4 = 0 \implies x = 1
   $$

   Значит, $x = 1$ — это экстремум.

3. **Проверим знак второй производной:**

   Чтобы определить, является ли эта точка точкой максимума, найдем вторую производную $f(x)$:

   $$
   f''(x) = \frac{d}{dx}(-4x + 4) = -4
   $$

   Вторая производная $f''(x) = -4$ отрицательна, следовательно, функция $f(x)$ имеет максимум в точке $x = 1$.

4. **Найдем значение функции в точке максимума:**

   Подставим $x = 1$ в исходную функцию:

   $$
   f(1) = -2(1)^2 + 4(1) + 1 = -2 + 4 + 1 = 3
   $$

Функция $f(x) = -2x^2 + 4x + 1$ имеет максимум в точке $x = 1$, при этом значение функции в этой точке равно $3$.

#### Пример поиска минимума

Найдите точку минимума функции $f(x) = x^2 - 4x + 3$ и определите значение функции в этой точке.

1. **Найдем производную функции:**

   Чтобы найти точку минимума, сначала вычислим первую производную функции $f(x)$:

   $$
   f'(x) = \frac{d}{dx}(x^2 - 4x + 3) = 2x - 4
   $$

2. **Найдем экстремумы:**

   Экстремумы — это значения $x$, при которых первая производная $f'(x)$ равна нулю:

   $$
   f'(x) = 2x - 4 = 0
   $$

   Решаем это уравнение:

   $$
   2x - 4 = 0 \implies x = 2
   $$

   Значит, $x = 2$ — это экстремум.

3. **Проверим знак второй производной:**

   Чтобы определить, является ли эта точка точкой минимума, найдем вторую производную $f(x)$:

   $$
   f''(x) = \frac{d}{dx}(2x - 4) = 2
   $$

   Вторая производная $f''(x) = 2$ положительна, следовательно, функция $f(x)$ имеет минимум в точке $x = 2$.

4. **Найдем значение функции в точке минимума:**

   Подставим $x = 2$ в исходную функцию:

   $$
   f(2) = (2)^2 - 4(2) + 3 = 4 - 8 + 3 = -1
   $$

Функция $f(x) = x^2 - 4x + 3$ имеет минимум в точке $x = 2$, при этом значение функции в этой точке равно $-1$.

## Интегральное исчисление

### Первообразная. Неопределенный интеграл

**Первообразная**

> Первообразной функции $F(x)$ для функции $f(x)$ называется такая функция $F(x)$, производная которой равна $f(x)$. То есть, $F'(x) = f(x)$. Проще говоря, первообразная — это функция, производная которой равна исходной функции.

Например, если $f(x) = 2x$, то $F(x) = x^2 + C$, где $C$ — произвольная константа. Это константа появляется, потому что производная любой константы равна нулю, и поэтому при нахождении первообразной мы не можем определить ее значение.

Существует бесконечное множество первообразных для одной и той же функции, отличающихся на произвольную константу $C$. Поэтому при записи первообразной всегда добавляют эту константу.

**Неопределенный интеграл**

Неопределенным интегралом функции $f(x)$ называется множество всех ее первообразных, и обозначается он следующим образом:

$$
\int f(x) \, dx = F(x) + C,
$$

где $F(x)$ — любая первообразная функции $f(x)$, а $C$ — произвольная постоянная.

Интеграл можно рассматривать как обратную операцию к дифференцированию. Если мы знаем производную функции, то можем восстановить саму функцию (с точностью до константы) с помощью интеграла.

**Свойства неопределенного интеграла**



- **Линейность интеграла**
    
    1. Интеграл от суммы функций равен сумме их интегралов:
$$\int [f(x) + g(x)] \, dx = \int f(x) \, dx + \int g(x) \, dx$$

    2. Интеграл от произведения функции на константу равен произведению этой константы на интеграл от функции:

$$\int k \cdot f(x) \, dx = k \cdot \int f(x) \, dx$$

- **Интеграл от производной**
    
$$\int F'(x) \, dx = F(x) + C$$

- **Интеграл от нуля**

$$\int 0 \, dx = C$$

#### Пример вычисления неопределенного интеграла

Рассмотрим интеграл функции $ \cos(x) + 2x $. Нам нужно найти неопределенный интеграл:

$$
\int (\cos(x) + 2x) \, dx
$$

Разделим этот интеграл на два отдельных (по свойству аддитивности):

$$
\int \cos(x) \, dx + \int 2x \, dx
$$

Первый интеграл - это стандартный интеграл из "таблицы интегралов":

$$
\int \cos(x) \, dx = \sin(x) + С
$$

Во втором интеграле можно вынести 2 из интеграла (по свойству линейности):

$$
\int 2x \, dx = 2 \int x \, dx
$$

Во втором интеграле мы перешли к табличному интегралу, поэтому можем его вычислить:

$$
2 \int x \, dx = 2 \cdot \frac{x^2}{2} + C = x^2 + C
$$

Теперь объединим результаты и добавим константу $C$. Обратите внимание, что при объединении результатов мы не добавляем $2C$, а добавляем просто $C$, так как разница между ними не принципиальна:

$$
\sin(x) + x^2 + C
$$

Итак, решение интеграла:

$$
\int (\cos(x) + 2x) \, dx = \sin(x) + x^2 + C
$$

### Продвинутые приемы интегрирования

При интегрировании сложных функций стандартные методы могут оказаться неэффективными. Для упрощения задачи применяются специальные приемы интегрирования, такие как *метод замены переменной* и *метод интегрирования по частям*, которые позволяют упростить исходный интеграл и найти его значение.

### Определенный интеграл

**Определенный интеграл и пределы интегрирования**

Определенный интеграл функции $f(x)$ на интервале $[a, b]$ обозначается следующим образом:

$$
\int_{a}^{b} f(x) \, dx
$$

Здесь:
  - $f(x)$ — функция, которую мы интегрируем.
  - $a$ и $b$ — пределы интегрирования, где $a$ называется нижним пределом, а $b$ — верхним пределом.
  - $dx$ указывает на переменную интегрирования и на то, что мы суммируем значения функции на бесконечно малых участках интервала.

Определенный интеграл рассматривается как сумма значений функции $f(x)$ на промежутке $[a, b]$, умноженных на бесконечно малую ширину этих участков. В результате мы получаем аккумулированное значение функции на данном интервале.

**Как вычисляется определенный интеграл?**

1. *Нахождение неопределенного интеграла (первообразной)*

   Сначала мы находим неопределённый интеграл (или первообразную) функции $f(x)$.

   Например, если $f(x) = x^2$, то неопределённый интеграл будет:

   $$
   \int x^2 \, dx = \frac{x^3}{3} + C
   $$

2. *Подстановка пределов интегрирования*

   После нахождения первообразной $F(x)$, мы можем вычислить определённый интеграл. Для этого подставляем сначала верхний предел $b$, а затем нижний предел $a$ в первообразную:

   $$
   \int_{a}^{b} f(x) \, dx = F(b) - F(a)
   $$

   Этот результат даёт значение интеграла, которое обычно интерпретируется как площадь под графиком функции $f(x)$ между $x = a$ и $x = b$.

**Геометрический смысл определенного интеграла**

Если функция $f(x)$ непрерывна и неотрицательна на интервале $[a, b]$, то определенный интеграл $\int_{a}^{b} f(x) \, dx$ равен *площади области*, ограниченной графиком функции $f(x)$, осью $Ox$ и прямыми $x = a$ и $x = b$.

**Свойства определенного интеграла**

- **Линейность**
    
1. Интеграл от суммы функций равен сумме их интегралов:

$$\int_{a}^{b} [f(x) + g(x)] \, dx = \int_{a}^{b} f(x) \, dx + \int_{a}^{b} g(x) \, dx$$
    
2. Интеграл от произведения функции на константу равен произведению этой константы на интеграл:
    
$$\int_{a}^{b} k \cdot f(x) \, dx = k \cdot \int_{a}^{b} f(x) \, dx$$

- **Свойство аддитивности**
    
    Это свойство позволяет разбить интеграл на сумму интегралов на подынтервалах.

$$\int_{a}^{b} f(x) \, dx + \int_{b}^{c} f(x) \, dx = \int_{a}^{c} f(x) \, dx$$
    

- **Изменение пределов интегрирования**
    
1. Если поменять местами пределы интегрирования, то интеграл поменяет знак:
    
$$\int_{a}^{b} f(x) \, dx = -\int_{b}^{a} f(x) \, dx$$
    
2. Если оба предела интегрирования совпадают, то значение интеграла равно нулю:
    
$$\int_{a}^{a} f(x) \, dx = 0$$

- **Интеграл от неотрицательной функции**

    Если $f(x) \geq 0$ на $[a, b]$, то и интеграл будет неотрицательным:

$$\int_{a}^{b} f(x) \, dx \geq 0$$

#### Пример вычисления определенного интеграла

Рассмотрим определенный интеграл:

$$
\int_{0}^{2} (x + 2) \, dx
$$

Решим его поэтапно. Сначала найдем неопределенный интеграл:

$$
\int (x + 2) \, dx = \int x \, dx + \int 2 \, dx
$$

Первый интеграл:

$$
\int x \, dx = \frac{x^2}{2} + С
$$

Второй интеграл:

$$
\int 2 \, dx = 2x + С
$$

Теперь объединим результаты. Обратите внимание, что при вычислении определенного интеграла мы можем не учитывать $C$ и просто его не писать:

$$
\frac{x^2}{2} + 2x
$$

Теперь вычислим определенный интеграл, подставив пределы от 0 до 2. Обозначается он таким образом:

$$
\left[ \frac{x^2}{2} + 2x \right]_{0}^{2}
$$

Сначала подставляем $x = 2$:

$$
\frac{2^2}{2} + 2(2) = \frac{4}{2} + 4 = 2 + 4 = 6
$$

Теперь подставляем $x = 0$:

$$
\frac{0^2}{2} + 2(0) = 0 + 0 = 0
$$

Теперь находим разность:

$$
6 - 0 = 6
$$

Итак, значение определенного интеграла:

$$
\int_{0}^{2} (x + 2) \, dx = 6
$$

### Интегралы на бесконечных промежутках

#### Определение интегралов на бесконечных промежутках

Когда мы говорим об интегралах на бесконечных промежутках, имеем в виду определенные интегралы, где один из пределов интегрирования (или оба) равны бесконечности. Такие интегралы называют несобственными, и их вычисление требует специального подхода.



**Интеграл от $a$ до $+\infty$**

Рассмотрим интеграл вида

$$
\int_{a}^{+\infty} f(x) \, dx
$$

Этот интеграл определяется как предел:

$$
\int_{a}^{+\infty} f(x) \, dx = \lim_{b \to +\infty} \int_{a}^{b} f(x) \, dx
$$

Здесь сначала вычисляется интеграл от $a$ до конечного значения $b$, а затем берется предел при $b \to +\infty$. Если предел существует и конечен, то интеграл считается сходящимся; в противном случае — расходящимся.



**Интеграл от $-\infty$ до $b$**

Аналогично, интеграл вида

$$
\int_{-\infty}^{b} f(x) \, dx
$$

определяется как предел:

$$
\int_{-\infty}^{b} f(x) \, dx = \lim_{a \to -\infty} \int_{a}^{b} f(x) \, dx
$$



**Интеграл от $-\infty$ до $+\infty$**

Если функция определена на всей числовой оси, то можно рассмотреть интеграл на всем интервале:

$$
\int_{-\infty}^{+\infty} f(x) \, dx
$$

Он определяется как сумма двух несобственных интегралов:

$$
\int_{-\infty}^{+\infty} f(x) \, dx = \int_{-\infty}^{c} f(x) \, dx + \int_{c}^{+\infty} f(x) \, dx
$$

где $c$ — произвольная точка на числовой оси.



#### Сходимость интегралов

Обратите внимание, что не все интегралы на бесконечных промежутках сходятся к каком-либо числу. На этом занятии мы не будем затрагивать этот момент, но в целом мы можем сказать, что сходимость несобственных интегралов на бесконечных промежутках зависит от поведения функции $f(x)$ на бесконечности.
