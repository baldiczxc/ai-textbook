### Основные термины и классическое определение вероятности

#### Событие

В теории вероятностей событие — это исход (результат) или набор исходов (результатов), которые могут произойти в итоге выполнения случайного эксперимента.

Примеры событий:
1. Выпадет "орёл" или "решка" после подбрасывания монеты
2. Выпадет 4 после броска игрального кубика

События могут быть нескольких видов:

1. *Достоверное событие*
    
    Это событие, которое обязательно произойдет. Например, при подбрасывании обычной монеты одно из событий "орёл" или "решка" обязательно произойдет.

2. *Невозможное событие*

    Это событие, которое не может произойти. Например, при броске обычного игрального кубика не может выпасть число 7.

3. *Случайное событие*

    Это событие, которое может произойти, а может и не произойти. Например, при подбрасывании монеты выпадет "орёл".

#### Пространство элементарных событий

Пространство элементарных событий — это множество всех возможных исходов случайного эксперимента. Каждый исход в этом множестве называется элементарным событием.

Пространство элементарных событий помогает нам увидеть все возможные исходы эксперимента, что является основой для вычисления вероятностей различных событий.

Примеры элементарных событий:

1. Подбрасывание монеты

    { "орёл", "решка" }.

2. Бросок игрального кубика:

    { 1, 2, 3, 4, 5, 6 }.

3. Подбрасывание двух монет:

    { (орёл, орёл), (орёл, решка), (решка, орёл), (решка, решка) }.

#### Классическое определение вероятности

Если эксперимент имеет $n$ равновозможных и взаимно исключающих исходов, и $m$ из них благоприятствуют событию $A$, то вероятность события $A$ определяется как отношение числа благоприятных исходов к общему числу исходов:

$$ P(A) = \frac{n}{m}$$​

Обратите внимание, что это формула применяется только при следующих условиях:
1. Каждый исход эксперимента имеет одинаковую вероятность произойти (все исходы равновозможны).
2. Не могут происходить одновременно (все исходы взаимно исключают друг друга).

#### Примеры



**Пример 1**

Какова вероятность выпадения орла, если подбросить монету?

- **Эксперимент**: Подбрасывание монеты.
- **Пространство элементарных событий**: { "орёл", "решка" }.
- **Событие $A$**: Выпадение "орла".
- **Количество благоприятных исходов $m$**: 1 (только "орёл").
- **Общее количество исходов $n$**: 2 (всего два исхода: "орёл" и "решка").

По классическому определению вероятности:

$$ P(A) = \frac{m}{n} = \frac{1}{2} $$

**Пример 2**

Если подбросить игральный шестигранный кубик, то с какой вероятностью выпадет четное число?

- **Эксперимент**: Бросок игрального кубика.
- **Пространство элементарных событий**: { 1, 2, 3, 4, 5, 6 }.
- **Событие $A$**: Выпадение четного числа.
- **Количество благоприятных исходов $m$**: 3 (четные числа: 2, 4, 6).
- **Общее количество исходов $n$**: 6 (всего шесть граней кубика).

По классическому определению вероятности:

$$ P(A) = \frac{m}{n} = \frac{3}{6} = \frac{1}{2} $$

### Совместные и несовместные события

#### Совместные события

Совместные события — это события, которые могут произойти одновременно. То есть, существует хотя бы один исход эксперимента, при котором оба события происходят.

К примеру, при подбрасывании кубика можно выделить два события:
1. Событие А: выпадет четное число (2, 4, 6).
2. Событие B: выпадет число больше 3 (4, 5, 6).

Числа 4 и 6 входят и в событие A, и в событие B. Значит, события A и B совместны, так как они могут происходить одновременно.

#### Несовместные события

Несовместные события (или взаимоисключающие события) — это события, которые не могут произойти одновременно. То есть, нет ни одного исхода эксперимента, при котором оба события происходят.



К примеру, при подбрасывании кубика можно выделить два события:
1. Событие A: Выпадет число 2.
2. Событие B: Выпадет число 5.

Число не может быть одновременно и 2, и 5. Значит, события A и B несовместны, так как они не могут происходить одновременно.

### Теорема сложения вероятностей несовместных событий

Вероятность появления одного из двух несовместных событий А или B (без разницы какого), равна сумме вероятностей этих событий:

$$ P(A + B) = P(A \cup B) = P(A) + P(B) $$

Аналогичный факт справедлив и для большего количества несовместных событий, например, для трёх несовместных событий:

$$ P(A + B + C) = P(A \cup B \cup C) = P(A) + P(B) + P(C) $$

#### Пример

**Задача**

В урне находятся 3 красных шара, 2 синих шара и 5 зелёных шаров. Из урны случайным образом вытаскивают один шар. Найти вероятность того, что вытянутый шар будет либо красного, либо синего цвета.

**Решение**

Сначала найдём вероятности отдельных событий:

1. **Событие $A$**: вытянут красный шар.
   - Количество красных шаров: 3.
   - Общее количество шаров: $ 3 + 2 + 5 = 10 $.
   - Вероятность $ P(A) = \frac{3}{10} $.

2. **Событие $ B $**: вытянут синий шар.
   - Количество синих шаров: 2.
   - Общее количество шаров: $ 3 + 2 + 5 = 10 $.
   - Вероятность $ P(B) = \frac{2}{10} $.

Поскольку события "вытянут красный шар" и "вытянут синий шар" являются несовместными (шар не может быть одновременно и красным, и синим), мы можем использовать правило сложения вероятностей для несовместных событий:

$$ P(A \cup B) = P(A) + P(B) $$

Подставим значения вероятностей:

$$  P(A \cup B) = \frac{3}{10} + \frac{2}{10} = \frac{5}{10} = 0.5 $$

Вероятность того, что вытянутый шар будет либо красного, либо синего цвета, равна 0.5 или 50%.

### Теорема сложения вероятностей совместных событий

Если два события, A и B, могут произойти одновременно (они совместные), то вероятность того, что произойдет одно из них, равна сумме их вероятностей минус вероятность их одновременного наступления:

$$ P(A + B) = P(A \cup B) = P(A) + P(B) - P(A \cap B) $$

#### Пример

Представьте, что вы бросаете стандартный шестигранный кубик. Пусть событие $ A $ — это выпадение четного числа, а событие $ B $ — это выпадение числа больше 3.

- Четные числа на кубике: 2, 4, 6. Значит, $ P(A) = \frac{3}{6} = 0.5 $.
- Числа больше 3: 4, 5, 6. Значит, $ P(B) = \frac{3}{6} = 0.5 $.
- Числа, которые одновременно четные и больше 3: 4, 6. Значит, $ P(A \cap B) = \frac{2}{6} = \frac{1}{3} \approx 0.333 $.

Теперь применим правило сложения вероятностей для совместных событий:

$$ P(A \cup B) = P(A) + P(B) - P(A \cap B) = 0.5 + 0.5 - 0.333 = 0.667 $$

Таким образом, вероятность того, что выпадет либо четное число, либо число больше 3 (или то и другое), составляет 0.667.

### Зависимые и независимые события

#### Независимые события

Независимые события — это события, которые не влияют друг на друга. Происхождение одного события никак не изменяет вероятность происхождения другого события.

Рассмотрим два события при подбрасывании двух монет:
1. Событие A: на первой монете выпадает "орёл".
2. Событие B: на второй монете выпадает "орёл".

Вероятность выпадения "орла" на первой монете никак не влияет на вероятность выпадения "орла" на второй монете. Эти события независимы.

#### Зависимые события

Зависимые события — это события, которые влияют друг на друга. Происхождение одного события изменяет вероятность происхождения другого события.

Рассмотрим два события при извлечении карт из колоды:
1. Событие A: Вытащить туз из полной колоды (52 карты).
2. Событие B: Вытащить короля после того, как туз был уже вытащен и не возвращён в колоду.

Если мы вытащили туз и не вернули его обратно в колоду, то количество оставшихся карт изменилось, что влияет на вероятность вытаскивания короля.

### Теорема умножения вероятностей независимых событий

Вероятность совместного появления независимых событий A и B равна произведению вероятностей этих событий:

$$ P(AB) = P(A \cap B) = P(A) \cdot P(B) $$


#### Пример

**Задача:**

Какова вероятность того, что при подбрасывании двух монет обе выпадут "орлом"?

**Решение:**

1. **Событие $ A $**: На первой монете выпадает "орёл".
   - Вероятность $ P(A) = \frac{1}{2} $.

2. **Событие $ B $**: На второй монете выпадает "орёл".
   - Вероятность $ P(B) = \frac{1}{2} $.

Так как события независимы, вероятность их совместного наступления:

$$ P(A \cap B) = P(A) \cdot P(B) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4} $$

### Теорема умножения вероятностей зависимых событий

#### Условная вероятность

Вероятность события B, вычисленная в предположении того, что событие A уже произошло, называется условной вероятностью наступления события B и обозначается через $$P(B | A)$$

При этом события A и B называют зависимыми событиями.

#### Теорема

Вероятность совместного появления двух зависимых событий равна произведению вероятности одного из них на условную вероятность другого, вычисленную в предположении, что первое событие уже произошло:

$$ P(AB) = A(A \cap B) = P(A) \cdot P(B | A)$$

#### Пример

**Задача:**

Какова вероятность того, что при вынимании двух карт подряд из полной колоды (без возвращения) сначала выпадет туз, а затем король?

**Решение:**

1. **Событие $ A $**: Вытащить туз.
   - Вероятность $ P(A) = \frac{4}{52} = \frac{1}{13} $.

2. **Событие $ B $**: Вытащить король после туза.
   - После того, как туз вытащен, остаётся 51 карта, из них 4 короля.
   - Условная вероятность $ P(B | A) = \frac{4}{51} $.

Вероятность совместного наступления зависимых событий:

$$ P(A \cap B) = P(A) \times P(B | A) = \frac{1}{13} \times \frac{4}{51} = \frac{4}{663} $$

### Теорема о полной вероятности

Теорема о полной вероятности помогает вычислять вероятность события, если мы знаем вероятность события в различных условиях, которые покрывают все возможные случаи.

#### Исчерпывающие события

Исчерпывающие события — это набор событий, который охватывает все возможные исходы данного эксперимента. Это означает, что одно из событий в таком наборе обязательно произойдёт.

#### Теорема

Предположим, что у нас есть несколько событий $ B_1, B_2, \ldots, B_n $, которые:
1. Являются взаимно исключающими (несовместными), то есть никакие два события не могут произойти одновременно.
2. Являются исчерпывающими, то есть совокупность этих событий охватывает все возможные исходы.

Если $ A $ — это произвольное событие, то вероятность события $ A $ можно найти с помощью следующей формулы:

$$ P(A) = \sum_{i=1}^{n} P(A \cap B_i) $$

Где $ P(A \cap B_i) $ — это вероятность того, что произошло событие $ A $ и одно из событий $ B_i $.





Эту формулу можно переписать с учетом условной вероятности:

$$ P(A) = \sum_{i=1}^{n} P(A | B_i) \cdot P(B_i) $$

где $ P(A | B_i) $ — это условная вероятность события $ A $, если известно, что произошло событие $ B_i $.

#### Пример

**Задача:**

В школе есть три класса: 5-А, 5-Б и 5-В. В 5-А учится 30 учеников, в 5-Б — 25 учеников, а в 5-В — 20 учеников. Из каждого класса проводится тестирование. Вероятность того, что ученик из 5-А получит хороший результат на тесте, равна 0.8, для 5-Б — 0.6, для 5-В — 0.5. Найти вероятность того, что случайно выбранный ученик из всех классов получит хороший результат на тесте.

**Решение:**

1. Определим вероятности выбора ученика из каждого класса:

   - Общее количество учеников: $ 30 + 25 + 20 = 75 $.
   - Вероятность выбора ученика из 5-А: $ P(B_1) = \frac{30}{75} = \frac{2}{5} $.
   - Вероятность выбора ученика из 5-Б: $ P(B_2) = \frac{25}{75} = \frac{1}{3} $.
   - Вероятность выбора ученика из 5-В: $ P(B_3) = \frac{20}{75} = \frac{4}{15} $.

2. Найдем условные вероятности получения хорошего результата на тесте:

   - Вероятность хорошего результата для 5-А: $ P(A | B_1) = 0.8 $.
   - Вероятность хорошего результата для 5-Б: $ P(A | B_2) = 0.6 $.
   - Вероятность хорошего результата для 5-В: $ P(A | B_3) = 0.5 $.

3. Применим теорему о полной вероятности:

    $$
    P(A) = P(A | B_1) \cdot P(B_1) + P(A | B_2) \cdot P(B_2) + P(A | B_3) \cdot P(B_3)
    $$

    Подставим значения и вычислим:

    $$
    P(A) = (0.8 \cdot \frac{2}{5}) + (0.6 \cdot \frac{1}{3}) + (0.5 \cdot \frac{4}{15})
    $$

    $$
    P(A) \approx 0.32 + 0.2 + 0.1333 = 0.6533
    $$

Вероятность того, что случайно выбранный ученик из всех классов получит хороший результат на тесте, составляет примерно 0.6533 или 65.33%.

### Формула Байеса

#### Априорная вероятность

Априорная вероятность — это тип вероятности, основанный на предположениях или знаниях, которые существуют до проведения какого-либо эксперимента или сбора новых данных. Этот вид вероятности определяется на основе логических рассуждений или прошлых знаний о ситуации, и обычно не требует эмпирических данных для её вычисления.

Вероятность выпадения орла или решки при броске честной монеты — классический пример априорной вероятности. Если мы предполагаем, что монета сбалансирована, мы можем априори определить, что вероятность выпадения орла равна 0,5.

#### Формула

Формула Байеса, также известная как теорема Байеса, обновлять вероятность события на основе новой информации. Она выражает условную вероятность одного события при условии, что другое событие уже произошло. Формула Байеса выглядит следующим образом:

$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$

где:
- $ P(A|B) $ — условная вероятность события A при условии, что произошло событие B (апостериорная вероятность).
- $ P(B|A) $ — условная вероятность события B при условии, что произошло событие A.
- $ P(A) $ — априорная вероятность события A.
- $ P(B) $ — априорная вероятность события B.

Формула Байеса помогает пересчитать вероятность события $A$, когда мы знаем, что произошло событие $B$. Она связывает условные вероятности событий и помогает обновлять наши знания по мере получения новой информации.

#### Пример

**Задача:**

Допустим, что 1% людей в популяции больны определенной болезнью. Тест на эту болезнь имеет следующие характеристики:
- Если человек болен, тест даёт положительный результат в 99% случаев (чувствительность теста).
- Если человек здоров, тест даёт положительный результат в 5% случаев (ложноположительный результат).

Какова вероятность того, что человек болен, если его тест дал положительный результат?

**Решение:**

1. **Обозначим события**:
   - $ A $ — человек болен.
   - $ B $ — тест дал положительный результат.

2. **Данные**:
   - $ P(A) = 0.01 $ — вероятность, что человек болен.
   - $ P(\neg A) = 0.99 $ — вероятность, что человек здоров.
   - $ P(B|A) = 0.99 $ — вероятность положительного теста, если человек болен.
   - $ P(B|\neg A) = 0.05 $ — вероятность положительного теста, если человек здоров.

3. **Найдём $ P(B) $** — общую вероятность положительного теста:
   $$ P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A) $$
   Подставим значения:
   $$ P(B) = 0.99 \cdot 0.01 + 0.05 \cdot 0.99 = 0.0099 + 0.0495 = 0.0594 $$

4. **Применим формулу Байеса**, чтобы найти вероятность того, что человек болен при положительном результате теста:
   $$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$
   Подставим значения:
   $$ P(A|B) = \frac{0.99 \cdot 0.01}{0.0594} = \frac{0.0099}{0.0594} \approx 0.1667 $$


Вероятность того, что человек действительно болен при положительном результате теста, составляет примерно 16.67%.
